ä¸ºä»€ä¹ˆæ‰€æœ‰å¯¹è±¡éœ€è¦æœ‰è¿™æ ·ä¸€ä¸ªæ–¹æ³•å‘¢ï¼Œéœ€è¦èƒ½è¿”å›è¿™æ ·ä¸€ä¸ªå“ˆå¸Œå€¼å‘¢ï¼Œå…¶å®æ˜¯ä¸ºäº†åœ¨å­—å…¸å’Œé›†åˆè¿™æ ·ä½¿ç”¨æ•£åˆ—è¡¨å­˜å‚¨æ•°æ®çš„ç»“æ„ä¸­æŸ¥æ‰¾æ•°æ®ï¼Œå› ä¸ºæ•£åˆ—è¡¨æŸ¥æ‰¾å€¼æ˜¯æ ¹æ®`key`çš„å“ˆå¸Œå€¼æ¥æŸ¥æ‰¾çš„ï¼Œå…·ä½“çš„å®ç°ç»†èŠ‚æŸ¥çœ‹ã€Šå­—å…¸åˆ°åº•æ˜¯æœ‰åºè¿˜æ˜¯æ— åºã€‹ä¸€èŠ‚ã€‚ç®€å•æ¥è¯´ï¼Œä»¥å­—å…¸ä¸ºä¾‹ï¼Œä¼šå…ˆæ ¹æ®è¦æŸ¥æ‰¾çš„`key`çš„å“ˆå¸Œå€¼çš„å…¶ä¸­å‡ ä½è®¡ç®—å‡ºä¸€ä¸ªåç½®ï¼Œç„¶åæ ¹æ®è¿™ä¸ªåç½®æ‰¾åˆ°æ•£åˆ—è¡¨ä¸­çš„ä¸€ä¸ªè¡¨å…ƒï¼Œå¦‚æœè¡¨å…ƒä¸ºç©ºï¼Œåˆ™æŠ›å‡º`KeyError`ï¼Œå¦‚æœä¸ä¸ºç©ºï¼Œä¼šæŸ¥çœ‹è¡¨å…ƒé‡Œé¢å­˜å‚¨çš„`key`å’Œè¦æŸ¥æ‰¾çš„`key`æ˜¯ä¸æ˜¯ä¸€è‡´ï¼Œä¸€è‡´åˆ™è¿”å›`value`ï¼Œå¦‚æœä¸åŒï¼Œåˆ™å‘ç”Ÿäº†æ•£åˆ—å†²çªï¼Œåˆ™ä¼šå†ä»è¦æŸ¥æ‰¾çš„`key`çš„å“ˆå¸Œå€¼é‡æ–°å†å–å‡ ä½è®¡ç®—åç½®ï¼Œå†é‡å¤ä»¥ä¸Šè¿‡ç¨‹ã€‚

````python
class C:
    def __eq__(self, value):
        if hash(self) == hash(value):
            return True

    def __hash__(self):
        return 123456


c1 = C()
c1.name = "c1"
c2 = C()
c2.name = "c2"

c1 == c2  
c1 is c2  # isæ˜¯æ ¹æ®idæ¥åˆ¤æ–­çš„ï¼Œæ‰€ä»¥c1 is not c2
````

````python
d = {c1: 42}

d[c1]
d[c2]
````

å¦‚ä¸Šçš„ä¾‹å­ï¼Œå³ä½¿`c1`å’Œ`c2`æ˜¯ä¸¤ä¸ªä¸åŒçš„å®ä¾‹ï¼Œä½†æ˜¯ç”±äºä»–ä»¬çš„å“ˆå¸Œå€¼æ˜¯ç›¸åŒçš„ï¼Œè€Œä¸”ç›¸ç­‰ï¼ˆé€šè¿‡å®šä¹‰`__eq__`æ–¹æ³•å®ç°ï¼‰ã€‚æ‰€ä»¥å¯ä»¥é€šè¿‡`c2`æ¥æŸ¥æ‰¾`c1`ã€‚        

æ‰€ä»¥ï¼Œä¸€ä¸ªä¸å¯å˜å¯¹è±¡ï¼Œå®ç°äº†`__eq__`æ–¹æ³•ï¼ŒåŒæ—¶éœ€è¦å®ç°`__hash__`æ–¹æ³•ï¼Œæ‰èƒ½ä½œä¸ºå­—å…¸çš„é”®ï¼Œå¹¶ä¸”è¦ä¿è¯åœ¨`a == b`ä¸ºçœŸçš„æƒ…å†µä¸‹`hash(a) == hash(b)`ä¹Ÿä¸ºçœŸï¼Œè¿™æ ·æ‰èƒ½å®ç°æ ¹æ®ç›¸åŒçš„å€¼ï¼Œæ¥å¯¹å­—å…¸è¿›è¡Œè¯»å–ã€‚

````python
t1 = tuple((1, 2))
t2 = tuple((1, 2))

d = {t1: "t1"}
d
d[t2]
````

è€Œä¸€ä¸ªå¯å˜å¯¹è±¡ï¼Œä¸æ˜¯è¯´ä¸èƒ½å®ç°`__hash__`æ–¹æ³•ï¼Œè€Œæ˜¯æ²¡æ³•å®ç°å“ˆå¸Œå€¼ä¹Ÿç›¸åŒçš„æ—¶å€™ï¼Œå€¼ä¸€å®šç›¸åŒï¼Œè¿™å°±ç ´åäº†ç°æœ‰çš„æŸ¥æ‰¾çš„åŸåˆ™ï¼š

````python
from collections import UserList


class MyList(UserList):
    def __hash__(self):
        return id(self)


mylist1 = MyList([1, 2, 3])
d = {mylist1: "I am mylist"}
d
````

````python
import traceback

mylist2 = MyList([1, 2, 3])
try:
    d[mylist2]
except KeyError:
    traceback.print_exc()
````

#### å…³äº`exec`

`exec()`å¯ä»¥æ‰§è¡Œå­—ç¬¦ä¸²å½¢å¼çš„è¯­å¥ï¼Œä½†æ˜¯è¦æ³¨æ„çš„æ˜¯å®ƒåœ¨å±€éƒ¨èŒƒå›´ä¸­çš„è¡Œä¸ºï¼Œä¸»è¦æœ‰ä¸¤ç‚¹ï¼š
1. åœ¨å±€éƒ¨èŒƒå›´å†…ï¼Œæ€»æ˜¯æ”¶é›†æ‰€æœ‰çš„å±€éƒ¨å˜é‡ï¼Œç»„æˆä¸€ä¸ªå­—å…¸ä¼ é€’ç»™`exec`ï¼Œä¹Ÿå°±æ˜¯è¯´ä¼ é€’ç»™`exec`çš„å¹¶ä¸æ˜¯å®é™…çš„å±€éƒ¨å˜é‡ï¼Œè€Œæ˜¯ä¸€ä¸ªå‰¯æœ¬ã€‚
2. æ¯æ¬¡è¿è¡Œ`locals()`æ–¹æ³•ï¼Œå°±ç›¸å½“äºé‡æ–°æ”¶é›†ä¸€æ¬¡æ‰€æœ‰çš„å±€éƒ¨å˜é‡ã€‚

````python
def func():
    a = 12
    exec("b = a + 2")
    print(b)

func()
````

å¦‚ä¸Šï¼Œåœ¨å‡½æ•°èŒƒå›´å†…ï¼Œå…¶å®å¹¶æ²¡æœ‰`b`å˜é‡ã€‚å¯ä»¥è¿™æ ·åšï¼š

````python
def func():
    a = 12
    loc = locals()
    exec("b = a + 2")
    b = loc['b']  # ä¸èƒ½ä½¿ç”¨locals()['b']ï¼Œå› ä¸ºæ­¤æ—¶è¿è¡Œlocals()ä¼šåˆä¸€æ¬¡æ”¶é›†å‡½æ•°çš„å±€éƒ¨å˜é‡æ›´æ–°å‰¯æœ¬ï¼Œå‡½æ•°çš„å±€éƒ¨å˜é‡æ²¡æœ‰b, ç›¸å½“äºè¦†ç›–æ‰ç°æœ‰å‰¯æœ¬
    print(b)

func()
````

æ³¨æ„ï¼Œ`locals()`è¿”å›çš„ä¹Ÿæ˜¯å±€éƒ¨èŒƒå›´å†…æ‰€æœ‰å˜é‡æ„æˆçš„å­—å…¸çš„å‰¯æœ¬ï¼Œå’Œä¼ å…¥`exec`çš„æ˜¯åŒä¸€ä¸ªï¼Œå¹¶ä¸ä¼šè¿è¡Œ`locals()`çš„æ—¶å€™äº§ç”Ÿä¸€ä¸ªå‰¯æœ¬ï¼Œæ‰§è¡Œ`exec`çš„æ—¶å€™åˆäº§ç”Ÿä¸€ä¸ªå‰¯æœ¬ã€‚åŒæ—¶è¦æ³¨æ„ï¼Œæ¯æ¬¡è¿è¡Œ`locals()`æ–¹æ³•ï¼Œä¼šæ”¶é›†å½“å‰çš„æ‰€æœ‰å±€éƒ¨å˜é‡æ›´æ–°è¿™ä¸ªå‰¯æœ¬ï¼ˆå§‹ç»ˆåªæœ‰ä¸€ä¸ªå‰¯æœ¬ï¼‰ï¼š

````python
def func():
    a = 12
    loc = locals()
    print(loc)
    exec("a += 2")
    print(loc)
    locals()
    print(loc)

func()
````

å¦‚æœè¿™æ ·çš„è¡Œä¸ºè®©ä½ ç–‘æƒ‘ï¼Œå¯ä»¥ä½¿ç”¨`exec`çš„åä¸¤ä¸ªå‚æ•°æ˜ç¡®çš„ä¼ é€’è¡¨ç¤ºä½œç”¨åŸŸçš„å­—å…¸ç»™å®ƒï¼Œ`exec`åªä¼šæ“ä½œè¿™ä¸ªå­—å…¸é‡Œçš„å˜é‡ï¼Œä¸ä¼šå½±å“åˆ°å®é™…çš„å˜é‡ï¼š

````python
def func():
    a = 12
    loc = {"a": a}
    exec("a += 2", {}, loc)
    print(loc['a'])
    print(a)
    
func()
````

#### withè¯­å¥

`with`æ¯”è¾ƒå¥½ç†è§£ï¼Œå°±æ˜¯åé¢æ¥ä¸€ä¸ªæœ‰`__enter__`å’Œ`__exit__`ç±»çš„å®ä¾‹ï¼Œä¸è¿‡æœ‰å‡ ç‚¹è¦æ³¨æ„ï¼š
1. å½“åœ¨`with`å†…éƒ¨ä»£ç å—é‡Œé¢æŠ›å‡ºå¼‚å¸¸ï¼Œä¼šä¼ é€’åˆ°`__exit__`ä¸­ï¼Œæ­¤æ—¶éœ€è¦è¿”å›`True`å‘Šè¯‰è§£é‡Šå™¨ï¼Œå¼‚å¸¸å·²ç»å¤„ç†ï¼Œå¦åˆ™å¼‚å¸¸ä¼šå‘ä¸Šå†’æ³¡ï¼š

````python
class C:
    def __enter__(self):
        pass
    def __exit__(self, exc_type, exc_value, traceback):
        if exc_type is not None:
            print(f"{exc_value}")

with C():
    raise TypeError("type error")
````

````python
class C:
    def __enter__(self):
        pass
    def __exit__(self, exc_type, exc_value, traceback):
        if exc_type is not None:
            print(f"{exc_value}")
            return True

with C():
    raise TypeError("type error")
````

2. `contextlib.context.manager`è£…é¥°å™¨è£…é¥°ä¸€ä¸ªç”Ÿæˆå™¨æ—¶ï¼Œæœ€ä½³å®è·µæ˜¯å°†`yield`è¯­å¥æ”¾åœ¨`try...except...finally`ä¸­é—´ï¼Œæ¥æ•è·`with`å†…éƒ¨æŠ›å‡ºçš„å¼‚å¸¸ï¼Œå¦‚æœè¦å‘ä¸Šå†’æ³¡ï¼Œéœ€è¦åœ¨`except`ä¸­ä¸»åŠ¨å°†å¼‚å¸¸å†æ¬¡æŠ›å‡ºã€‚

````python
from contextlib import contextmanager

@contextmanager
def with_demo():
    try:
        yield
    except Exception as e:
        print(e)
        
with C():
    raise TypeError("type error")
````

````python
from contextlib import contextmanager

@contextmanager
def with_demo():
    try:
        yield
    except Exception as e:
        print(e)
        raise e
        
with with_demo():
    raise TypeError("type error")
````

ä½†æ˜¯ä¸€äº›è¾ƒå°çš„æ•´æ•°ï¼Œpythonå†…éƒ¨è¿›è¡Œäº†ä¼˜åŒ–ï¼Œä»–åœ¨å†…å­˜ä¸­å§‹ç»ˆä¿å­˜ç€ä¸€äº›è¾ƒå°çš„æ•´æ•°ï¼Œå½“æœ‰å˜é‡èµ‹å€¼æ—¶ï¼Œç›´æ¥æŠŠè¿™äº›å°çš„æ•´æ•°çš„åœ°å€èµ‹å€¼ç»™å˜é‡ï¼š

````python
z = []
for i in range(-1000, 1000):
    for j in range(-1000, 1000):
        if id(i) == id(j):
            z.append(i)
print([z[0], z[-1]])
````

````python
class C:
    def __init__(self, name):
        self.name = name
        
    def __format__(self, format_spec):
        if format_spec == "obj":
            return "The obj's name is: {.name}".format(self)
        else:
            return "{!r}".format(self)

c = C("C")
"{:obj}".format(c)
"{}".format(c)
````

### å…¶å®ƒçš„æ•°æ®ç»“æ„

é™¤äº†å¸¸ç”¨çš„`dict`,`list`ï¼Œ`set`ï¼Œ`python`è¿˜æœ‰å¾ˆå¤šå…¶å®ƒçš„æ•°æ®ç»“æ„ï¼Œå¯èƒ½ç”±äºå‰é¢å‡ ä¸ªå¤ªå¥½ç”¨äº†ï¼Œå…¶å®ƒçš„ä½¿ç”¨å¾—å°±æ²¡é‚£ä¹ˆå¤šï¼Œç”šè‡³çŸ¥é“çš„äººéƒ½ä¸å¤šï¼Œè¿™é‡Œå¯¹æ‰€æœ‰å†…ç½®çš„æ•°æ®ç»“æ„è¿›è¡Œä¸€ä¸ªæ¢³ç†ï¼Œå¹¶å°è¯•ç»™å‡ºä½¿ç”¨åœºæ™¯ã€‚

#### `queue`æ¨¡å—

`queue`æ¨¡å—ä¸‹ä¸»è¦åŒ…å«ä¸‰ç§é˜Ÿåˆ—æ•°æ®ç»“æ„ï¼Œåˆ†åˆ«æ˜¯æ™®é€šå…ˆè¿›å…ˆå‡ºçš„`Queue`ï¼Œåè¿›å…ˆå‡ºçš„`LiFoQueue`å’Œä¸€ä¸ªä¼˜å…ˆçº§é˜Ÿåˆ—`PriorityQueue`ï¼Œå¦å¤–è¿˜æœ‰ä¸€ä¸ªæ— ç•Œçš„`SimpleQueue`ï¼Œä¹Ÿæ˜¯å…ˆè¿›å…ˆå‡ºé˜Ÿä¼ï¼Œä½†æ˜¯ç¼ºå°‘ä»»åŠ¡è·Ÿè¸ªç­‰é«˜çº§åŠŸèƒ½ã€‚è¿™äº›é˜Ÿåˆ—éƒ½æ˜¯çº¿ç¨‹å®‰å…¨çš„ï¼Œæ‰€ä»¥å¤šçº¿ç¨‹éœ€è¦ä½¿ç”¨å…±äº«æ•°æ®æ—¶ï¼Œä¼˜å…ˆé€‰æ‹©ä½¿ç”¨è¿™å‡ ä¸ªé˜Ÿåˆ—ã€‚  

è¿™é‡Œä¸»è¦è®°å½•å¦‚ä½•åœ¨å¤šçº¿ç¨‹ä¸­ä½¿ç”¨é˜Ÿåˆ—ï¼Œå…ˆä»‹ç»å‡ ä¸ª`queue`çš„ä¸»è¦æ–¹æ³•å’Œå±æ€§ï¼š
- `put(item, block=True, timeout=None)`ï¼šå‘é˜Ÿåˆ—æ”¾å…¥ä¸€ä¸ªä»»åŠ¡ï¼Œå¦‚æœ`block`ä¸º`True`ï¼Œæ„æ€æ˜¯é˜»å¡`put`æ‰€åœ¨çš„å½“å‰çº¿ç¨‹ï¼Œç­‰ç€é˜Ÿåˆ—ç©ºå‡ºæ§½ä½ï¼Œå¦‚æœ`timeout`ä¸ºNoneï¼Œå°±ä¼šä¸€ç›´ç­‰ä¸‹å»ã€‚å¦‚æœè®¾ç½®äº†å€¼ï¼Œé‚£ä¹ˆåœ¨ç­‰å¾…æ—¶é—´åˆ°è¾¾ä»¥åï¼Œé˜Ÿåˆ—è¿˜æ²¡æœ‰ç©ºå‡ºæ§½ä½ï¼Œåˆ™æŠ›å‡º`full`çš„é”™è¯¯ã€‚å¦‚æœ`block`ä¸º`False`ï¼Œé‚£ä¹ˆå°±ä¸ä¼šç­‰å¾…ï¼Œç«‹å³å‘é˜Ÿåˆ—æ’å…¥ï¼Œå¦‚æœé˜Ÿåˆ—æ˜¯æ»¡çš„ï¼Œç›´æ¥æŠ›å‡º`full`é”™è¯¯ã€‚è¿™ä¸ªæ—¶å€™`timeout`å‚æ•°å°±å¤±æ•ˆäº†ã€‚
- `put_nowait(item)`ï¼šå°±ç›¸å½“äº`block`ä¸º`False`ã€‚
- `get(block=True, timeout=None)`ï¼šå’Œ`put`æ­£å¥½ç›¸åï¼Œä¸€ä¸ªæ˜¯æ”¾ï¼Œä¸€ä¸ªæ˜¯å–ï¼ŒåŒæ—¶æŠ›å‡ºçš„æ˜¯`empty`é”™è¯¯ã€‚
- `get_nowait()`ï¼šç›¸å½“äº`get`ä¸­çš„`block`ä¸º`False`ã€‚
- `task_done()`ï¼šå½“çº¿ç¨‹`get`ä¸€ä¸ªä»»åŠ¡ï¼Œå¹¶ä¸”è¿™ä¸ªä»»åŠ¡æ‰§è¡Œå®Œäº†ï¼Œéœ€è¦è°ƒç”¨é˜Ÿåˆ—çš„`task_done`æ–¹æ³•ï¼Œé€šçŸ¥é˜Ÿåˆ—ä»»åŠ¡å·²ç»æ‰§è¡Œå®Œæ¯•ï¼Œæ­¤æ—¶é˜Ÿåˆ—æ‰ä¼šé‡Šæ”¾ç©ºçš„æ§½ä½ã€‚æ‰€ä»¥ï¼Œè¿™ä¸ªæ–¹æ³•è°ƒç”¨å’Œä»»åŠ¡åº”è¯¥æ˜¯ä¸€å¯¹ä¸€çš„å…³ç³»ï¼Œå¦‚æœè°ƒç”¨æ¬¡æ•°å¤§äºä»»åŠ¡çš„æ•°é‡ï¼Œä¼šæŠ›å‡º`ValueError`çš„é”™è¯¯ã€‚
- `join()`ï¼šé˜»å¡ä¸»çº¿ç¨‹ï¼Œä¸€ç›´åˆ°é˜Ÿåˆ—é‡Œçš„æ‰€æœ‰ä»»åŠ¡éƒ½æ‰§è¡Œå®Œæ¯•ã€‚

å…·ä½“çœ‹ä¾‹å­ï¼š

````python
from queue import Queue
from threading import Thread
from time import sleep

# å®šä¹‰ä¸€ä¸ªå­˜æ”¾ä»»åŠ¡çš„é˜Ÿåˆ—
q = Queue()


def do_job(name, job_number):
    print(f"{name} start do his job {job_number}")
    sleep(job_number)
    print(f"{name} finish his job {job_number}")


# å®šä¹‰ä¸€ä¸ªåœ¨çº¿ç¨‹é‡Œé¢æ‰§è¡Œçš„å‡½æ•°
def worker():
    # å¾ªç¯è·å–ä»»åŠ¡
    while True:
        # ä»é˜Ÿåˆ—é‡Œé¢è·å–ä»»åŠ¡ï¼Œå› ä¸ºblockä¸ºTrueï¼Œå¦‚æœæ­¤æ—¶é˜Ÿåˆ—ä¸ºç©ºï¼Œçº¿ç¨‹ä¼šé˜»å¡ï¼Œç­‰å¾…é˜Ÿåˆ—é‡Œé¢æœ‰ä»»åŠ¡
        task = q.get(block=True)
        # å½“æ”¶åˆ°çš„ä»»åŠ¡æ˜¯Noneçš„æ—¶å€™ï¼Œè·³å‡ºå¾ªç¯
        if task is None:
            print("thread end")
            break
        worker_name, job_number = task
        # è¿™é‡Œç›¸å½“äºå®šä¹‰å®é™…è¦åšçš„å·¥ä½œ
        do_job(worker_name, job_number)
        # å®Œæˆä»¥åè¦è°ƒç”¨é˜Ÿåˆ—çš„task_done()æ–¹æ³•ï¼Œé€šçŸ¥é˜Ÿåˆ—ä»»åŠ¡å·²ç»å®Œæˆ
        q.task_done()


ts = []  # åˆ›å»ºä¸€ä¸ªçº¿ç¨‹çš„ç©ºåˆ—è¡¨

# åˆ›å»º3ä¸ªçº¿ç¨‹ï¼Œå¹¶å¯åŠ¨ï¼Œå¯åŠ¨å®Œäº†ä¼šé©¬ä¸Šå›åˆ°ä¸»çº¿ç¨‹ï¼Œå°†çº¿ç¨‹æ·»åŠ åˆ°åˆ—è¡¨
for i in range(6):
    t = Thread(target=worker)
    # å¯åŠ¨çº¿ç¨‹ï¼Œå› ä¸ºä»»åŠ¡é˜Ÿåˆ—ç°åœ¨æ˜¯ç©ºçš„ï¼Œæ‰€ä»¥çº¿ç¨‹ä¼šåœ¨q.getå¤„é˜»å¡ï¼Œç­‰å¾…é˜Ÿåˆ—é‡Œæœ‰ä»»åŠ¡
    t.start()
    ts.append(t)

# æ‰€æœ‰çš„ä»»åŠ¡
tasks = [("Tom", 3), ("John", 2), ("Tom", 2), ("John", 1), ("Tom", 1), ("John", 3)]

# å°†ä»»åŠ¡putè¿›é˜Ÿåˆ—ï¼Œæ­¤æ—¶é˜»å¡çš„çº¿ç¨‹ä¼šå¼€å§‹æ¥æ”¶åˆ°ä»»åŠ¡ï¼Œå‘ä¸‹æ‰§è¡Œ
for t in tasks:
    q.put(t)

# é˜Ÿåˆ—éœ€è¦joinï¼Œç­‰å¾…æ‰€æœ‰ä»»åŠ¡æ‰§è¡Œå®Œæ¯•
q.join()

# æ‰€æœ‰ä»»åŠ¡æ‰§è¡Œå®Œæ¯•ï¼Œæ­¤æ—¶é˜Ÿåˆ—åˆä¸ºç©ºï¼Œæ‰€æœ‰çº¿ç¨‹åˆä¼šé˜»å¡åœ¨q.get()ï¼Œæ‰€ä»¥éœ€è¦å‘é˜Ÿåˆ—putä¸ä»»åŠ¡ç›¸åŒæ•°é‡çš„Noneï¼Œä»¥ä¾¿çº¿ç¨‹ç»“æŸ
for i in range(len(tasks)):
    q.put(None)

# é»˜è®¤ä¼šç­‰å¾…æ‰€æœ‰çº¿ç¨‹ç»“æŸï¼Œä¹Ÿå¯ä»¥æ‰‹åŠ¨é˜»å¡ï¼Œç­‰å¾…çº¿ç¨‹ç»“æŸ
for t in ts:
    t.join()
````

bytearrayåŸºæœ¬ä¸Šå’Œbytesæ˜¯ä¸€æ ·çš„ï¼Œä¸è¿‡å®ƒå¯ä»¥åŸåœ°è¿›è¡Œä¿®æ”¹ï¼Œæ¯”å¦‚ï¼š

````python
b = bytearray(b'hello')
# æ³¨æ„ï¼Œbytearrayå¯ä»¥æ ¹æ®ç´¢å¼•ç›´æ¥èµ‹å€¼ï¼Œä½†æ˜¯åªèƒ½èµ‹å€¼ä¸ºæ•´æ•°ï¼Œä¸èƒ½b[1]=b'a'ï¼Œb'a'æ˜¯ä¸€ä¸ªå­—èŠ‚å­—ç¬¦ä¸²å¯¹è±¡ï¼Œæ­¤æ—¶ä¼šæŠ¥é”™
b[1] = 97
b
# ä½†æ˜¯æ³¨æ„ï¼Œå¦‚æœæ˜¯åˆ‡ç‰‡çš„è¯ï¼Œå¯ä»¥ç›´æ¥èµ‹å€¼å­—èŠ‚å­—ç¬¦ä¸²ï¼Œä¸ºä»€ä¹ˆä¼šè¿™æ ·ï¼Œè¿˜æ²¡ç†è§£æ¸…æ¥š
b[1:2] = b'c'
b
````

bytearrayä¸»è¦æ˜¯å¯ä»¥ç”¨ä½œè¯»å–äºŒè¿›åˆ¶æ•°æ®æ—¶çš„ç¼“å†²åŒºï¼Œå…·ä½“ä¾‹å­å¯è§ã€Špython cookbooxã€‹5.8èŠ‚ã€‚

#### bytearrayå’Œmemoryview

ä¹‹å‰ä¸€ç›´æœªç†è§£æœ‰äº†bytearrayï¼Œä¸ºä»€ä¹ˆè¿˜è¦memoryviewã€‚å¯¹äºæ”¯æŒç¼“å†²åŒºåè®®çš„å¯¹è±¡ï¼Œç›´æ¥readinto bytearrayä¸å°±å¥½äº†å—ï¼Ÿæœ€è¿‘æ‰å‘ç°bytearrayå’Œmemoryviewçš„ä¸åŒï¼š

````python
a = bytearray(b'hello world')
b = a[:2]  # bytearrayçš„åˆ‡ç‰‡å¹¶ä¸”èµ‹å€¼ç»™ä¸€ä¸ªå˜é‡ï¼Œæ­¤æ—¶åˆ›å»ºäº†ä¸€ä¸ªå‰¯æœ¬
b[:] = b'bb'
a  # ä¿®æ”¹è¿™ä¸ªå‰¯æœ¬ï¼Œaä¸å˜
m = memoryview(a)[:2]  # åˆ›å»ºä¸€ä¸ªå†…å­˜è§†å›¾ï¼Œå…¶åˆ‡ç‰‡æ˜¯ä¸€ä¸ªæ˜ å°„
m.tobytes()  # memoryviewæ˜¯ä¸€ä¸ªå¯¹è±¡ï¼Œéœ€è¦è°ƒç”¨tobytesæ–¹æ³•æ‰èƒ½çœ‹åˆ°å†…å®¹
m[:] = b'mm' # ä¿®æ”¹è¿™ä¸ªåˆ‡ç‰‡ï¼Œaä¹Ÿå‘ç”Ÿäº†å˜åŒ–
a
````

å¯è§ï¼Œå¯¹äºmemoryviewåˆ‡ç‰‡åšçš„ä»»ä½•èµ‹å€¼æ“ä½œéƒ½ä¼šä¿®æ”¹åŸbytearrayã€‚æ³¨æ„ï¼Œä¸èƒ½ä¿®æ”¹memoryviewåˆ‡ç‰‡çš„å¤§å°ï¼Œå› ä¸ºå®ƒåªæ˜¯å†…å­˜ä¸­å›ºå®šå¤§å°çš„æ˜ å°„ã€‚

````python
try:
    m[:] = b'mmm'
except ValueError as e:
    print(e)
````

å¸¸è§çš„ç”¨é€”æ¯”å¦‚ä¸€ä¸ªå¥—æ¥å­—ï¼Œæ¯æ¬¡æ¥æ”¶å›ºå®šçš„å¤§å°ï¼Œå°±å¯ä»¥å…ˆåˆ›å»ºä¸€ä¸ªå®Œæ•´å¤§å°çš„bytearrayï¼Œç„¶åç”¨memoryviewåŒ…è£…ï¼Œæ¯æ¬¡æ¥æ”¶åˆ°å›ºå®šå¤§å°çš„å­—èŠ‚ï¼Œå°±å¯ä»¥ç›´æ¥èµ‹å€¼ç»™memoryviewåˆ‡ç‰‡ï¼Œè¿™æ ·å°±é¿å…äº†å¤§é‡çš„å†…å­˜çš„å¤åˆ¶å’Œè®¡ç®—ï¼Œæé«˜æ€§èƒ½ã€‚

`time`åœ¨å®˜æ–¹æ–‡æ¡£ä¸­,è¢«åˆ†ç±»åœ¨é€šç”¨æ“ä½œç³»ç»ŸæœåŠ¡ä¸­ï¼Œ`datetime`æ˜¯å¯¹`time`çš„å°è£…ï¼Œåˆ†ç±»åœ¨æ•°æ®ç±»å‹ä¸­ï¼Œä»åˆ†ç±»å°±å¯ä»¥çœ‹å‡º,`time`å€¾å‘äºæä¾›å¼€å‘ä¸­ä¸ç³»ç»Ÿæ—¶é—´ç›¸å…³çš„åŠŸèƒ½ï¼Œå®ƒè¿˜åŒ…å«è¯¸å¦‚ è¿”å›æŒ‡å®šçš„çº¿ç¨‹idçš„ç‰¹å®šäºçº¿ç¨‹çš„CPUæ—¶é—´æ—¶é’Ÿçš„clk_id è¿™æ ·çš„å‡½æ•°ï¼Œè€Œ`datetime`æ›´å€¾å‘äºç”¨ä½œæ•°æ®åˆ†æå’Œå¤„ç†ã€‚

### `time`æ¨¡å—

åœ¨`python`ä¸­ï¼Œæ‰€æœ‰çš„æ—¶é—´æœ‰ä¸‰ç§åŸºæœ¬æ ¼å¼ï¼Œåˆ†åˆ«æ˜¯æ—¶é—´æˆ³ï¼Œæ—¶é—´å…ƒç»„å’Œæ—¶é—´å­—ç¬¦ä¸²ã€‚å®ƒä»¬çš„å…³ç³»å¦‚ä¸‹ï¼š


è¿™é‡Œè®°å½•å‡ ä¸ªè¦ç‚¹ï¼š
1. æ‰€æœ‰è¿™äº›å‡½æ•°éƒ½æ˜¯`time`çš„é¡¶å±‚å‡½æ•°ï¼Œæ˜¯æ²¡æœ‰å¯¹è±¡çš„è½¬æ¢æ–¹æ³•çš„ã€‚
2. `localtime`æ˜¯è½¬æ¢ä¸ºæœ¬åœ°æ—¶é—´ï¼Œ`gmtime`æ˜¯è½¬æ¢ä¸ºæ ‡å‡†æ—¶é—´ã€‚
3. `asctime`å’Œ`ctime`ç›¸å½“äºä¸¤ä¸ªå¿«æ·æ–¹æ³•ï¼Œç›´æ¥å°†æ—¶é—´å…ƒç»„å’Œæ—¶é—´æˆ³è½¬æ¢ä¸ºå›ºå®šæ ¼å¼çš„å­—ç¬¦ä¸²ã€‚
4. æ ¼å¼åŒ–å‚æ•°è§[å®˜æ–¹æ–‡æ¡£](https://docs.python.org/zh-cn/3/library/time.html)ã€‚

````python
import time

now = time.time()
localtime = time.localtime(now)
gmtime = time.gmtime(now)
localtime
gmtime
time.strftime("%Y-%m-%d %H:%M", localtime)
time.ctime(now)
time.asctime(gmtime)
````

#### æ—¶åŒºè½¬æ¢

`datetime`æ¨¡å—å¯ä»¥æŠŠutcæ ¼å¼çš„æ—¶é—´è½¬æ¢æˆæœ¬åœ°æ—¶é—´ï¼Œä½†æ˜¯ä¸å¤ªå¥½ç”¨ã€‚å…ˆæ¥çœ‹ä¸ªæ¦‚å¿µï¼Œ`python`çš„`datetime`å¯¹è±¡å’Œ`time`å¯¹è±¡åˆ†ä¸ºæ„ŸçŸ¥å‹å’Œç®€å•å‹ä¸¤ç§ï¼Œç®€å•æ¥è¯´ï¼Œä¸å¸¦æ—¶åŒºä¿¡æ¯çš„æ˜¯ç®€å•å‹ï¼Œå¸¦äº†æ—¶åŒºä¿¡æ¯çš„æ˜¯æ„ŸçŸ¥å‹ï¼Œå‡†ç¡®çš„åˆ¤æ–­æ¡ä»¶æ˜¯ï¼Œå¯¹è±¡åŒæ—¶æ»¡è¶³ä»¥ä¸‹ä¸¤ä¸ªæ¡ä»¶åˆ™ä¸ºæ„ŸçŸ¥å‹ï¼š
1. `d.tzinfo` ä¸ä¸º `None`
2. `d.tzinfo.utcoffset(d)` ä¸è¿”å› `None`

å…ˆçœ‹ä»¥ä¸‹ä¸€æ®µä»£ç ï¼Œä½¿ç”¨`astimezone`æ–¹æ³•å°†`utc`æ ‡å‡†æ—¶é—´è½¬æ¢æˆæœ¬åœ°æ—¶é—´ï¼š

````python
from datetime import datetime, timezone

now = datetime.now()
now_utc = now.replace(tzinfo=timezone.utc)
local = now_utc.astimezone()

print(now)
print(now_utc)
print(local)
````

å¦‚æœä¸äº†è§£æ„ŸçŸ¥å‹å’Œç®€å•å‹ä¸¤ç§æ—¶é—´ç±»å‹ï¼Œä¼šè§‰å¾—è½¬æ¢æ–¹æ³•å¾ˆåˆ«æ‰­ã€‚å› ä¸ºå¹³æ—¶æˆ‘ä»¬é€šè¿‡`datetime`çš„`now`å‡½æ•°è™½ç„¶è·å–çš„æ˜¯æœ¬æœºçš„æ—¶é—´ï¼ˆä¹Ÿå°±æ˜¯æœ¬åœ°çš„å½“å‰æ—¶é—´ï¼‰ï¼Œä½†æ˜¯æ˜¯ç®€å•å‹ï¼Œä¸å¸¦æ—¶åŒºä¿¡æ¯ã€‚å½“ä½¿ç”¨`replace`æ–¹æ³•åŠ å…¥æ—¶åŒºä¿¡æ¯çš„æ—¶å€™ï¼Œå˜æˆäº†æ„ŸçŸ¥å‹ï¼Œä½†æ˜¯æ³¨æ„ï¼Œæ­¤æ—¶åªæ˜¯ç®€å•åŠ äº†æ—¶åŒºä¿¡æ¯ï¼Œå¹¶æœªå‘ç”Ÿæ—¶åŒºçš„è½¬æ¢ã€‚å†ä½¿ç”¨`astimezone`å‡½æ•°å°†`utc`æ—¶é—´è½¬æ¢æˆæœ¬åœ°æ—¶é—´ã€‚

å¦‚æœè¦åœ¨æ—¶åŒºä¹‹é—´è½¬æ¢ï¼Œæœ€ç®€å•çš„æ–¹æ³•æ˜¯ä½¿ç”¨`pytz`åº“ï¼ˆã€Šç¼–å†™é«˜è´¨é‡pythonçš„59ä¸ªæ–¹æ³•ã€‹ä¸­è¯´è¦æ‰§è¡Œæ—¶åŒºåç§»çš„æ“ä½œï¼Œéœ€è¦å…ˆè½¬æˆ`utc`æ ¼å¼çš„`datetime`å¯¹è±¡ï¼Œä½†æ˜¯è‡ªå·±æµ‹è¯•ä»…ä»…åªæ˜¯æ—¶åŒºè½¬æ¢çš„è¯ï¼Œè²Œä¼¼å¯ä»¥ç›´æ¥è½¬æ¢ï¼‰ï¼š

````python
import pytz

now = datetime.now()
tz_sh = pytz.timezone('Asia/Shanghai')
sh_dt = tz_sh.localize(now)
tz_eastern = pytz.timezone('US/Eastern')
eastern_dt = tz_eastern.normalize(sh_dt)
print(sh_dt)
print(eastern_dt)
````

#### `rf`æ­é…ä½¿ç”¨æ³¨æ„äº‹é¡¹

#### `search`ã€`findall`ï¼Œ`finditer`å’Œçš„åŒºåˆ«

åˆšå¼€å§‹çš„æ—¶å€™å¯¹äºä»€ä¹ˆæ—¶å€™ç”¨å“ªä¸ªæ–¹æ³•æ€»æ˜¯å¼„ä¸å¤ªæ¸…æ¥šï¼Œæ‰€ä»¥è¿™é‡Œåšä¸ªæ¢³ç†ï¼Œå¹¶ä¸”æ€»ç»“é‡éš¾ç‚¹ï¼š
1. `findall`å’Œ`finditer`æ˜¯å…¨å±€åŒ¹é…ï¼Œç®€å•è¯´å¦‚æœæ–‡æœ¬ä¸­æœ‰2ä¸ªåŠä»¥ä¸Šèƒ½åŒ¹é…ä¸Šæ­£åˆ™ï¼Œåˆ™éƒ½ä¼šæ‰¾å‡ºæ¥ã€‚è€Œ`search`åªä¼šè¿”å›ç¬¬ä¸€ä¸ªåŒ¹é…æˆåŠŸçš„ç»“æœã€‚
2. å¦‚æœæ­£åˆ™è¡¨è¾¾å¼ä¸åŒ…å«åˆ†ç»„ï¼Œåˆ™`findall`è¿”å›æ‰€æœ‰åŒ¹é…æˆåŠŸçš„ç»“æœæ„æˆçš„åˆ—è¡¨ï¼Œå¦‚æœåŒ…å«åˆ†ç»„ï¼Œåˆ™è¿”å›çš„åˆ—è¡¨åªåŒ…å«æ‰€æœ‰åŒ¹é…æˆåŠŸçš„åˆ†ç»„æ„æˆçš„å…ƒç»„ï¼Œè€Œä¸ä¼šåŒ…å«åŒ¹é…æˆåŠŸçš„å®Œæ•´ç»“æœã€‚
3. `finditer`å¼¥è¡¥äº†æœ‰åˆ†ç»„æ—¶ï¼Œæ— æ³•è¿”å›å®Œæ•´ç»“æœçš„ç¼ºç‚¹ã€‚å®ƒæ˜¯ä¸€ä¸ªç”Ÿæˆå™¨ï¼Œæ¯æ¬¡è¿”å›ä¸€ä¸ªåŒ¹é…æˆåŠŸçš„`match`å¯¹è±¡ï¼Œå’Œ`search`è¿”å›çš„ç»“æœä¸€æ ·ã€‚å› æ­¤å¯ä»¥æ ¹æ®éœ€æ±‚æå–æƒ³è¦çš„ç»“æœã€‚

````python
import re
text = "This is a test string, That has two test lines"
m = re.search(r"(Th\w+).*?(test)", text)  # æ³¨æ„æ˜¯éè´ªå©ªåŒ¹é…

# searchæ–¹æ³•åªèƒ½åŒ¹é…ç¬¬ä¸€å¥ï¼ŒåŒ¹é…æˆåŠŸå°±è¿”å›äº†ã€‚
m.groups()
m.group(0)
````

````python
# æ­£åˆ™é‡Œé¢ä¸åŒ…å«åˆ†ç»„ï¼Œåˆ™è¿”å›å®Œæ•´çš„åŒ¹é…æˆåŠŸçš„ç»“æœ
re.findall(r"Th\w+.*?test", text)
````

````python
# æ­£åˆ™é‡Œé¢åŒ…å«åˆ†ç»„ï¼Œåˆ™åˆ—è¡¨åªè¿”å›åŒ¹é…æˆåŠŸçš„åˆ†ç»„æ„æˆçš„å…ƒç»„
re.findall(r"(Th\w+).*?(test)", text)
````

````python
# å¦‚æœæ­£åˆ™é‡Œé¢åŒ…å«åˆ†ç»„ï¼Œåˆè¦è·å–å®Œæ•´çš„åŒ¹é…ç»“æœï¼Œåªèƒ½ç”¨finditerï¼Œè¿”å›çš„æ¯ä¸€ä¸ªåŒ¹é…æˆåŠŸçš„matchå¯¹è±¡ï¼Œé€šè¿‡matchå¯¹è±¡çš„group(0)è·å–å®Œæ•´ç»“æœ
for m in re.finditer(r"(Th\w+).*?(test)", text):
    print(m.group(0))
````

#### æ‰©å±•è¡¨ç¤ºæ³•

è¿™é‡Œè®°å½•æ­£åˆ™è¡¨è¾¾å¼ä¸­å¸¸ç”¨çš„æ‰©å±•è¡¨ç¤ºæ³•ï¼š

```python
(?:)              :    ä¸æ•è·åˆ†ç»„
(?#)              :    ä¸åšåŒ¹é…ï¼Œåªæ˜¯ç”¨ä½œæ³¨é‡Š
(?=)              :    å³ç­‰äºé›¶å®½æ–­è¨€
(?!)              :    å³ä¸ç­‰äºé›¶å®½æ–­è¨€
(?<=)             :    å·¦ç­‰äºé›¶å®½æ–­è¨€
(?<!)             :    å·¦ä¸ç­‰äºé›¶å®½æ–­è¨€
(?(1)y|x)         :    æ ¹æ®æ•è·çš„åˆ†ç»„é€‰æ‹©ï¼Œå¦‚æœæ•è·çš„åˆ†ç»„1å­˜åœ¨ï¼Œå°±é€‰yï¼Œå¦åˆ™é€‰x
(?P<name>)        :    ä¸ºåˆ†ç»„å‘½å
 ```

#### åµŒå¥—çš„æ‹¬å·åˆ†ç»„

å¦‚æœåˆ†ç»„æ˜¯åµŒå¥—çš„ï¼Œåˆ™å…ˆä»å¤–å‘å†…æ•°ï¼Œæ•°å­—ä¾æ¬¡å¢åŠ ï¼Œç„¶åä»å·¦å¾€å³æ•°ï¼Œæ•°å­—ä¾æ¬¡å¢åŠ ï¼Œå¦‚ï¼š

````python
import re

text = "This is a test string"
m = re.search(r"(T\w+(is)).*(s(tr)\w+)", text)
m.groups()
````

#### å‘½ååˆ†ç»„ä»¥åŠå¦‚ä½•å¼•ç”¨åˆ†ç»„

å¯ä»¥å¯¹åˆ†ç»„å‘½åï¼Œä¸»è¦æ³¨æ„ä¸€ä¸‹å†™æ³•ï¼Œå‘½ååˆ†ç»„å¯ä»¥é€šè¿‡`group(name)`è°ƒç”¨ï¼ŒåŒæ—¶ä½ç½®åˆ†ç»„ä»ç„¶å¯ç”¨ã€‚å¦å¤–åç»­å¼•ç”¨æœ‰å‡ ç‚¹è¦æ³¨æ„ï¼š
1. åç»­å¼•ç”¨å‰é¢åŒ¹é…çš„ç»“æœï¼Œå¯ä»¥`(?P=)`è¿™æ ·å†™ï¼Œæ³¨æ„æ­¤æ—¶ä¸ä¼šæ•è·åç»­å¼•ç”¨çš„åˆ†ç»„ï¼Œå¦‚æœæƒ³è¦æ•è·ï¼Œè¦åœ¨å¤–é¢å†åŠ ä¸ªæ‹¬å·ã€‚
2. åœ¨`sub`æ–¹æ³•ä¸­å¦‚æœè¦å¼•ç”¨å‰é¢çš„å‘½ååˆ†ç»„ï¼Œä½¿ç”¨`\g<>`çš„æ–¹å¼ï¼Œå¦‚æœæ˜¯ä½ç½®åˆ†ç»„ï¼Œä½¿ç”¨`\1`,`\2`çš„å½¢å¼å¼•ç”¨ã€‚

````python
import re

text = "This is a test string\nThis has two lines"
m = re.search(r"(?P<a>t\w+s).*((?P=a)).*", text,
              re.I + re.S)  # è¦æ•è·å¼•ç”¨çš„åˆ†ç»„ï¼Œè¦å†åŠ ä¸ªæ‹¬å·ï¼Œå³((?P=a))ï¼Œå…¶ä¸­(?P=a)æ˜¯å¼•ç”¨å‰é¢çš„å‘½åæ•è·
m.group("a")
m.group(1)  # ä½ç½®åˆ†ç»„å’Œå‘½ååˆ†ç»„å‡å¯ç”¨
m.groups()
````

æ³¨æ„ï¼Œæ­¤æ—¶`r"(?P<a>t\w+s).*"`åŒ¹é…ä¸Šæ•´ä¸ªå¥å­ï¼Œç”¨`"\g<a> is over"`æ›¿æ¢æ‰æ•´ä¸ªå¥å­ï¼Œä¸”ä½¿ç”¨`\g<a>`ä¿ç•™äº†ä¹‹å‰åŒ¹é…ä¸Šçš„å‘½ååˆ†ç»„ã€‚æ­¤æ—¶ä½ç½®åˆ†ç»„ç”¨`\1,\2`æ¥è¡¨ç¤ºï¼š

````python
re.sub(r"(?P<a>t\w+s).*", r"\g<a> is over", text, flags=re.I + re.S)
````

#### é›¶å®½æ–­è¨€

æ‰€è°“é›¶å®½ï¼ŒæŒ‡çš„æ˜¯åŒ¹é…çš„æ˜¯ä½ç½®ï¼Œè€Œä¸æ˜¯å­—ç¬¦ï¼Œå®½åº¦ä¸º0ï¼Œå› æ­¤å«é›¶å®½ã€‚ç½‘ä¸Šæˆ–è€…ä¸€äº›ä¹¦é‡Œé¢éƒ½ç§°ä½œå‘å‰ï¼ˆå‘åï¼‰æ­£å‘ï¼ˆè´Ÿå‘ï¼‰é›¶å®½æ–­è¨€ï¼Œæ„Ÿè§‰å¾ˆä¸å¥½ç†è§£ï¼Œä¸å¦‚å«åšå·¦è¾¹ï¼ˆå³è¾¹ï¼‰ç­‰äºï¼ˆä¸ç­‰äºï¼‰é›¶å®½æ–­è¨€å¥½è®°ã€‚ä¸€èˆ¬çš„å½¢å¼æœ‰4ç§ï¼š
- `(?=æ¡ä»¶)`  è¿™ä¸ªä½ç½®çš„å³è¾¹ç­‰äºæ¡ä»¶ã€‚
- `(?!æ¡ä»¶)`  è¿™ä¸ªä½ç½®çš„å³è¾¹ä¸ç­‰äºæ¡ä»¶ã€‚
- `(?<=æ¡ä»¶)` è¿™ä¸ªä½ç½®çš„å·¦è¾¹ç­‰äºæ¡ä»¶ã€‚
- `(?<!æ¡ä»¶)` è¿™ä¸ªä½ç½®çš„å·¦è¾¹ä¸ç­‰äºæ¡ä»¶ã€‚

å¦‚ä¸‹çš„ä¾‹å­ï¼Œä»”ç»†åˆ†è¾¨å…¶ä¸­çš„ç»†å¾®å·®åˆ«ï¼š

````python
import re

text = "port up up states"
# æ‹¬å·é‡Œé¢åŒ¹é…ä¸€ä¸ªä½ç½®å’Œä¸€ä¸ªä»»æ„å­—ç¬¦ï¼Œè¡¨ç¤ºä»»æ„å­—ç¬¦å‰é¢çš„ä½ç½®åé¢éƒ½ä¸èƒ½æ˜¯up\s+upï¼Œå› æ­¤åŒ¹é…ä¸ä¸Šï¼Œè¿”å›None
re.search(r"^((?!up\s+up).)*$", text)
# ä¸‹é¢éƒ½å¯ä»¥åŒ¹é…ä¸Š
re.search(r"((?!up\s+up).)*$", text)
# ä»å¤´å¼€å§‹åŒ¹é…ï¼Œå½“åŒ¹é…åˆ°ç¬¬ä¸€ä¸ªuæ—¶ä¸æ»¡è¶³æ¡ä»¶ï¼Œå› æ­¤ç»“æœä¸º`port `
re.search(r"^((?!up\s+up).)*", text)
# ä¸åŠ æ‹¬å·ä»…ä»…è¡¨ç¤ºåªè¦å¼€å¤´ä¸æ˜¯`up\s+up`ï¼Œå°±å¯ä»¥åŒ¹é…åé¢æ‰€æœ‰å­—ç¬¦
re.search(r"^(?!up\s+up).*", text)
````

#### `groups`ï¼Œ`group`çš„åŒºåˆ«

æ­£åˆ™åŒ¹é…ä»¥åï¼Œå¦‚æœæ²¡æœ‰åŒ¹é…ä¸Šï¼Œè¿”å›`None`ï¼Œå¦‚æœåŒ¹é…ä¸Šï¼Œè¿”å›ä¸€ä¸ª`match`å¯¹è±¡ã€‚`match`å¯¹è±¡æœ‰`groups`å’Œ`group`æ–¹æ³•ï¼ŒåŒºåˆ«å¦‚ä¸‹ï¼š
1. `groups`è¿”å›æ•è·çš„åˆ†ç»„æ„æˆçš„å…ƒç»„ï¼Œå¦‚æœæ²¡æœ‰åˆ†ç»„ï¼Œåˆ™è¿”å›ä¸€ä¸ªç©ºçš„å…ƒç»„ã€‚
2. `group`æŒ‡å•ç‹¬çš„æ•è·çš„åˆ†ç»„ï¼Œå…¶ä¸­`group(0)`å’Œ`group()`å‡æŒ‡åŒ¹é…ä¸Šçš„æ•´ä¸ªå­—ç¬¦ä¸²ï¼Œæ³¨æ„ï¼Œæ•è·çš„åˆ†ç»„ä»1å¼€å§‹è®¡æ•°ã€‚

````python
import re

text = "This is a test string\nText has two lines"
m = re.search(r"(t\w+s).*(t\w+t).*", text, re.I + re.S)
m.groups()
m.group()
m.group(0)
m.group(1)
m.group(2)
````

æœ‰æ„æ€çš„æ˜¯ï¼Œä¸Šé¢çš„ç¨‹åºå¦‚æœæ˜¯è„šæœ¬ï¼Œå¯ä»¥æ­£å¸¸è¿è¡Œï¼Œä½†å¦‚æœæ˜¯å‘½ä»¤è¡Œæ¨¡å¼ï¼ŒåŒæ ·ä¼šæŠ›å‡º`SyntaxError: no binding for nonlocal 'var' found`é”™è¯¯ï¼Œæ¨æµ‹æ˜¯å› ä¸ºè„šæœ¬å’Œå‘½ä»¤è¡Œæ¨¡å¼ç¼–è¯‘çš„é¡ºåºä¸ä¸€æ ·ï¼Œå¯¹äºè„šæœ¬è€Œè¨€ï¼Œä¼šå°†å‡½æ•°ä½œä¸ºä¸€ä¸ªæ•´ä½“ç”±å¤–å‘å†…çš„è¿›è¡Œç¼–è¯‘ï¼Œå³å…ˆå£°æ˜å¤–å±‚å‡½æ•°çš„æ‰€æœ‰å˜é‡ï¼Œå†é€å±‚å£°æ˜å†…éƒ¨å‡½æ•°ä½“å†…çš„å˜é‡ï¼Œä»è€Œå¯ä»¥é¡ºåˆ©å°†`var`ä¸å¤–å±‚å‡½æ•°çš„å˜é‡è¿›è¡Œç»‘å®šï¼Œè€Œå‘½ä»¤è¡Œæ˜¯é€è¡Œç¼–è¯‘ï¼Œå½“è¿è¡Œåˆ°`nonlocal var`è¯­å¥æ—¶ï¼Œå¤–å±‚å‡½æ•°è¿˜æ²¡æœ‰å£°æ˜`var`å˜é‡ï¼Œå› æ­¤ä¼šæŠ¥é”™ã€‚æ‰€ä»¥ï¼Œåœ¨å‘½ä»¤è¡Œä¸­ï¼Œ`var = 42`è¯­å¥éœ€è¦æ”¾åœ¨`def printvar()`å‡½æ•°å®šä¹‰çš„å‰é¢ã€‚

æ­¤æ—¶ï¼Œè„šæœ¬æ¨¡å¼å’Œå‘½ä»¤è¡Œæ¨¡å¼éƒ½å¯ä»¥é¡ºåˆ©è¿è¡Œï¼Œå› ä¸ºæ²¡æœ‰`nonlocal`å…³é”®å­—ï¼Œå‘½ä»¤è¡Œæ¨¡å¼å¹¶ä¸ä¼šå»æ£€æŸ¥`printvar`å†…éƒ¨çš„`var`å˜é‡ä¸è°ç»‘å®šï¼Œå› æ­¤ä¸ä¼šæŠ¥é”™ã€‚
åœ¨è¿™é‡Œï¼Œä¸å£°æ˜`nonlocal`ä¼¼ä¹æ›´å¥½ï¼Œé‚£ä¹ˆï¼Œä»€ä¹ˆæƒ…å†µä¸‹éœ€è¦ä½¿ç”¨`nonlocal`å£°æ˜å‘¢ï¼Ÿç­”æ¡ˆæ˜¯ï¼Œå½“å†…å±‚å‡½æ•°éœ€è¦ä¿®æ”¹å¤–å±‚å‡½æ•°çš„å˜é‡æ—¶ï¼Œå¿…é¡»è¦ä½¿ç”¨`nonlocal`å£°æ˜ã€‚å¦‚ä¸‹ï¼š

````python
def sort_priority(numbers, group):
    found = False
    def helper(x):
        if x in group:
            found = True
            return (0, x)
        return (1, x)
    numbers.sort(key=helper)
    return found

numbers = [2, 5, 1, 8, 10]
sort_priority(numbers, [6, 7, 8, 9, 10])
print(numbers)
````

ä¸Šé¢çš„ä»£ç æ˜¯ç»™å‡ºä¸€ä¸ªç»„ï¼Œå¦‚æœ`numbers`å†…çš„å…ƒç´ åœ¨`group`ä¸­ï¼Œåˆ™ä¼˜å…ˆæ’åºã€‚è¿”å›çš„`numbers`æ’åºæ˜¯å¯¹çš„ï¼Œä½†æ˜¯`found`å€¼ä¸å¯¹ï¼Œå› ä¸ºè¿™é‡Œè¿”å›çš„å…¶å®æ˜¯å¤–å±‚å‡½æ•°çš„`found`ï¼Œå†…å±‚å‡½æ•°é€šè¿‡èµ‹å€¼ï¼Œåˆ›å»ºäº†ä¸€ä¸ªå±äºå†…å±‚å‡½æ•°çš„`found`å˜é‡ã€‚è€Œå†…å±‚å‡½æ•°çš„æœ¬æ„æ˜¯ä¿®æ”¹å¤–å±‚å‡½æ•°çš„`found`å˜é‡ï¼Œå› æ­¤è¿™é‡Œå¿…é¡»è¦ä½¿ç”¨`nonlocal`å£°æ˜ï¼Œå¦‚ä¸‹ï¼š

````python
def sort_priority(numbers, group):
    found = False
    def helper(x):
        nonlocal found
        if x in group:
            found = True
            return (0, x)
        return (1, x)
    numbers.sort(key=helper)
    return found

numbers = [2, 5, 1, 8, 10]
sort_priority(numbers, [6, 7, 8, 9, 10])
print(numbers)
````

### å¦‚ä½•ä¿®æ”¹å‡½æ•°ç­¾å

ä¿®æ”¹å‡½æ•°ç­¾åå¯ä»¥ä½¿ç”¨inspectæ¨¡å—ï¼Œå¦‚ä¸‹ï¼š

````python
import inspect

def add(a, b):
    return a + b

sig = inspect.signature(add)
params = list(sig.parameters.values())
# åœ¨sig.parameters.values()é‡Œæ·»åŠ æˆ–è€…åˆ é™¤å³å¯
params.append(inspect.Parameter('c', inspect.Parameter.POSITIONAL_OR_KEYWORD))
# sig.replaceè¿”å›ä¸€ä¸ªæ–°çš„ç­¾åï¼ŒåŸæ¥çš„ç­¾åä¸å˜
new_sig = sig.replace(parameters=params)
# sig.parametersæ˜¯ä¸€ä¸ªmappingproxyï¼Œä¿®æ”¹äº†parameters.values()ï¼Œåˆ™parametersä¹Ÿå‘ç”Ÿäº†å˜åŒ–
print(new_sig.parameters)
add.__signature__ = new_sig
````

````python
class Base:
    def __init__(self):
        print("Base inited.")


class ChildA(Base):
    def __init__(self):
        Base.__init__(self)
        print("ChildA inited")


class ChildB(Base):
    def __init__(self):
        Base.__init__(self)
        print("ChildB inited")


class User(ChildA, ChildB):
    def __init__(self):
        super().__init__()


# Userçš„super()è¿”å›çš„æ˜¯ChildAï¼ŒChildAä¸­ç›´æ¥è°ƒç”¨Baseï¼Œå› æ­¤ä¸ä¼šè°ƒç”¨ChildBä¸­çš„__init__æ–¹æ³•
u = User()
````

2. `super()`æ€»æ˜¯å°½å¯èƒ½çš„å°†æ­£ç¡®çš„å‚æ•°ä¼ é€’ç»™æ–¹æ³•ï¼Œåœ¨å†…éƒ¨åšäº†å¾ˆå¤šå·¥ä½œï¼Œä½†æ˜¯è¿™æ ·æœ‰æ—¶å€™åè€Œå®¹æ˜“è®©äººç–‘æƒ‘ã€‚

````python
class C:
    data = "data"

    def method(*args):
        print(args)

    @classmethod
    def cmethod(*args):
        print(args)

    @staticmethod
    def smethod(*args):
        print(args)


class SubC(C):
    def __init__(self):
        super().method()
        super().cmethod()
        super().smethod()
        print("=" * 50)
        super(SubC, self).method()  # æ­£ç¡®ï¼Œå°†selfä¼ é€’ç»™æ–¹æ³•
        super(SubC, self).cmethod()  # æ­£ç¡®ï¼Œè™½ç„¶è°ƒç”¨çš„æ˜¯super(SubC, self)ï¼Œä½†æ˜¯ä¼ å…¥æ–¹æ³•çš„ä¸æ˜¯selfè€Œæ˜¯class
        super(SubC, self).smethod()  # æ­£ç¡®ï¼Œæ²¡æœ‰å‚æ•°ä¼ é€’
        print("=" * 50)
        super(SubC, SubC).method()  # é”™è¯¯ï¼Œæ²¡æœ‰å‚æ•°ä¼ é€’ç»™æ–¹æ³•
        super(SubC, SubC).cmethod()  # æ­£ç¡®ï¼Œå°†classä¼ é€’ç»™æ–¹æ³•
        super(SubC, SubC).smethod()  # æ­£ç¡®ï¼Œæ²¡æœ‰å‚æ•°ä¼ é€’


sc = SubC()
````

ç”±æ­¤å¯è§ï¼Œ`super`å¹¶ä¸æ˜¯ç®€å•çš„å°†ç¬¬äºŒä¸ªå‚æ•°ä¼ é€’ç»™æ–¹æ³•ï¼Œè€Œæ˜¯åœ¨å†…éƒ¨è¿›è¡Œäº†åˆ¤æ–­ï¼Œå°½å¯èƒ½çš„å°†æ­£ç¡®çš„å¯¹è±¡ä¼ é€’ç»™æ–¹æ³•ï¼Œå”¯ä¸€çš„ä¾‹å¤–æ˜¯ç”¨`super(type, type2)`æ–¹å¼è°ƒç”¨å®ä¾‹æ–¹æ³•çš„æ—¶å€™ï¼Œç”±äº`super()`ä¸çŸ¥é“è¦ä¼ é€’ç»™æ–¹æ³•çš„å®ä¾‹æ˜¯ä»€ä¹ˆï¼Œæ­¤æ—¶ä¼šä»€ä¹ˆéƒ½ä¸ä¼ ã€‚å…¶å®ƒæƒ…å†µä¸‹ï¼Œéƒ½èƒ½ä¼ é€’æ­£ç¡®çš„å¯¹è±¡ç»™æ–¹æ³•ã€‚  

ä»”ç»†æ€è€ƒå¦‚ä¸‹ä»£ç ï¼ŒåŠ æ·±ç†è§£ï¼š

````python
from inspect import signature


class Base:
    def func(self):
        print(self.x + self.y)


class C(Base):
    def func(self):
        self.x = 2
        self.y = 3
        super(C, self).func()


c = C()
c.func()
````

#### å­ç±»æ‰©å±•property

ã€Špython cookbookã€‹å­ç±»æ‰©å±•propertyä¸­ï¼Œåœ¨å­ç±»ä¸­è·å–çˆ¶ç±»çš„ç‰¹æ€§ä½¿ç”¨`super().name`ï¼Œè€Œä¿®æ”¹ä½¿ç”¨çš„æ˜¯`super(C, C).name.__set__(self, value)`ï¼Œå§‹ç»ˆæ²¡å¤ªæ˜ç™½ä¿®æ”¹ä¸ºä»€ä¹ˆä¸èƒ½ç›´æ¥`super().name = value`ï¼Œæµ‹è¯•å¦‚ä¸‹ï¼š 

````python
class C:
    cname = "classname"
    
    def __init__(self, name):
        self._name = name
        
    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self, value):
        self._name = value

class SubC(C):
    def func(self, value):
        print(super()) # <super: <class 'SubC'>, <SubC object>>
        print(super(SubC, SubC)) # <super: <class 'SubC'>, <SubC object>>
        print(super(SubC, SubC).name)  # <property object at 0x000001D59FEADC78>
        print(super(SubC, SubC).cname)  # classname
        print(super().name)  # shl
        print(super().cname)  # classname
        super().cname = value  # æŠ›å‡ºAttributeError: 'super' object has no attribute 'cname'é”™è¯¯
        super().name = value
````

superçš„å†…éƒ¨æœºåˆ¶è¿˜æ˜¯æ²¡æœ‰å®Œå…¨å¼„æ˜ç™½ï¼Œæ ¹æ®ä»£ç çŒœæµ‹ï¼Œsuper()ä¼šè¿”å›ä¸€ä¸ªsuperç±»çš„å®ä¾‹ï¼Œå¯¹è¿™ä¸ªsuperå®ä¾‹çš„å±æ€§è¿›è¡ŒæŸ¥æ‰¾ï¼Œç›¸å½“äºåœ¨çˆ¶ç±»ï¼ˆä¾‹å­ä¸­çš„Cï¼‰ä¸­æŸ¥æ‰¾å±æ€§ï¼ŒçŒœæµ‹åœ¨å†…éƒ¨superå¤åˆ¶äº†SubCçš„MROåˆ—è¡¨ï¼Ÿå¹¶ä¸”å¦‚æœå±æ€§æ˜¯æ–¹æ³•çš„è¯ï¼Œä¼šä¼ é€’æ­£ç¡®çš„å‚æ•°ï¼ˆselfæˆ–è€…clsï¼‰ç»™æ–¹æ³•ã€‚æŸ¥æ‰¾ä»…ä»…é’ˆå¯¹è¯»æ“ä½œï¼Œä»»ä½•å†™æ“ä½œä¼šç›´æ¥ç»™superå®ä¾‹è¿›è¡Œèµ‹å€¼ï¼Œä¸ä¼šåœ¨MROä¸­æŸ¥æ‰¾å±æ€§ï¼Œå› æ­¤ä¼šå‡ºç°`'super' object has no attribute 'cname'`çš„é”™è¯¯ï¼Œæƒ³æƒ³ä¹Ÿå¯¹ï¼Œä¸å¯èƒ½åœ¨å­ç±»ä¸­ä¿®æ”¹çˆ¶ç±»çš„å±æ€§æˆ–è€…æ–¹æ³•ã€‚

å†çœ‹é—®é¢˜ï¼Œå¦‚æœç»§æ‰¿çš„å±æ€§æ˜¯çˆ¶ç±»çš„propertyæˆ–è€…æè¿°ç¬¦ï¼Œå­ç±»ä¸­super().nameç›¸å½“äºè°ƒç”¨`C.name.__get__(self, SubC)`ï¼Œæ³¨æ„ï¼Œselfæ˜¯å­ç±»çš„å®ä¾‹ï¼Œè€Œsuper(SubC, SubC)ç›¸å½“äº`C.name.__get__(None, SubC)`ï¼Œçˆ¶ç±»çš„nameæ˜¯ä¸€ä¸ªæè¿°ç¬¦ï¼Œè°ƒç”¨å®ƒå®é™…ä¸Šç›¸å½“äºè°ƒç”¨å®ƒçš„__get__æ–¹æ³•ã€‚å› æ­¤super().nameä¼ é€’äº†å­ç±»çš„å®ä¾‹ï¼Œè¿”å›å­—ç¬¦ä¸²ï¼Œè€Œ`super(SubC, SubC)`ç”±äºæ ¹æœ¬æ²¡æœ‰ä¸çŸ¥é“å­ç±»çš„å®ä¾‹æ˜¯å•¥ï¼Œå› æ­¤å‘__get__æ–¹æ³•ä¼ é€’æ˜¯Noneï¼Œå› æ­¤è¿”å›çš„æ˜¯æè¿°ç¬¦æœ¬èº«ã€‚

### `__new__`æ–¹æ³•ä»¥åŠåœ¨`__new__`æ–¹æ³•ä¸­ä½¿ç”¨`super()`çš„ä¸€ä¸ªbug

`__new__`æ–¹æ³•æœ‰3ç‚¹éœ€è¦æ³¨æ„ï¼š

2. å½“`super()`æ–¹æ³•çœç•¥å‚æ•°çš„æ—¶å€™ï¼Œ`super()`æ–¹æ³•ä¼šåœ¨å †æ ˆä¸­å¯»æ‰¾ç±»(`__class__`)ä»¥åŠç¬¬ä¸€ä¸ªå‚æ•°ã€‚

3. å¦‚æœåœ¨`__new__`ä¸­ä½¿ç”¨`super()`å¹¶ä¸”çœç•¥å‚æ•°ï¼Œå¿…é¡»æ˜¾ç¤ºçš„ä¼ å…¥ç¬¬ä¸€ä¸ª`cls`å‚æ•°ï¼Œä¸èƒ½ç›´æ¥ä½¿ç”¨åŠ¨æ€å‚æ•°æ¥ä¼ é€’`cls`(æ¯”å¦‚`*args`)ï¼Œå¦åˆ™ä¼šæŠ¥é”™ï¼ˆå¯ä»¥è§†ä¸ºä¸€ä¸ªbugï¼‰ã€‚æˆ–è€…æ˜¾ç¤ºçš„ä½¿ç”¨`super(type, obj_or_type)`æ–¹å¼è¿›è¡Œè°ƒç”¨ã€‚

````python
class C:
    def __new__(*args):
        print("Create c object.")
        # æ³¨æ„ä¸èƒ½ return super().__new__(*args)ï¼Œè¿™æ ·ä¼šæŠ¥é”™
        # æ³¨æ„ï¼Œè¿™ç§å½¢å¼argsä¸èƒ½åŒ…å«å…¶å®ƒå‚æ•°ï¼Œå› ä¸ºsuper(C, C)å…¶å®è°ƒç”¨çš„æ˜¯object.__new__(cls)ï¼Œåªæ¥æ”¶ä¸€ä¸ªclså‚æ•°ã€‚
        return super(C, C).__new__(*args)
    
#     # ä¹Ÿå¯ä»¥æ˜¯ä¸‹é¢è¿™ç§å½¢å¼
#     def __new__(cls, *args):
#         print("Create c object")
#         # å¦‚æœç»§æ‰¿äº†å…¶å®ƒçš„ç±»ï¼Œåˆ™è¦æ ¹æ®çˆ¶ç±»çš„æƒ…å†µä¼ é€’å‚æ•°
#         return super().__new__(cls)

c = C()
````

````python
class C:
    @staticmethod
    def sfunc():
        print("I am staticmethod.")
    
    def ifunc(self):
        print("in func:")
        self.sfunc()
        C.sfunc()
        print("end")

c = C()
c.ifunc()
````

````python
from copy import copy


class Bus:
    def __init__(self, passengers):
        self.passengers = [] if passengers is None else list(passengers)  # å¯¹ä¼ å…¥çš„åˆ—è¡¨è¿›è¡Œå¤åˆ¶

    def pick(self, passenger):
        self.passengers.append(passenger)

    def drop(self, passenger):
        self.passengers.remove(passenger)


bus1 = Bus(['Alice', 'John', 'Tom'])
bus2 = copy(bus1)
bus1.passengers.append('Alex')
bus1.passengers
bus2.passengers
````

### è®©å¯¹è±¡æ”¯æŒä¸Šä¸‹æ–‡åè®®

````python
class C:
    def __init__(self, name):
        self.name = name
        
    def __enter__(self):
        print("in __enter__")
        return self
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("in __exit__")
        return True
    
    
with C("telecomshy") as c:
    print(c.name)
````

### `StopIteration`å¼‚å¸¸

å•ç‹¬æŠŠ`StopIteration`å¼‚å¸¸æ‹¿å‡ºæ¥è®²ï¼Œä¸»è¦æ˜¯å› ä¸ºè¿™ä¸ªå¼‚å¸¸å’Œå…¶å®ƒå¼‚å¸¸ç›¸æ¯”ï¼Œæœ‰ä¸€ä¸ªä¸åŒçš„è€Œä¸”æ˜¯ç›¸å½“é‡è¦çš„åº”ç”¨åœºæ™¯ï¼Œå½“ç”Ÿæˆå™¨ï¼Œä¹Ÿå°±æ˜¯åç¨‹ç»“æŸçš„æ—¶å€™ï¼Œä¼šè‡ªåŠ¨æŠ›å‡º`StopIteration`å¼‚å¸¸ï¼Œæ­¤æ—¶ï¼Œ`StopIteration`ä¼šæ•è·åç¨‹æœ€åç»“æŸæ—¶çš„è¿”å›å€¼ï¼ŒæŠŠè¿”å›å€¼ä½œä¸ºå‚æ•°ï¼ˆä¸‹é¢ä¾‹å­ä¸­çš„`args`ï¼‰ï¼Œä¼ é€’ç»™`StopIteration`ï¼š

````python
def c():
    yield
    return 42

g = c()
try:
    next(g)
    next(g)
except StopIteration as e:
    print(e.args)
    print(e.value)
````

````python
s = StopIteration(1, 2, 3)
s.args
s.value
````

å½“åœ¨ç”Ÿæˆå™¨ä¸­returnï¼Œä¼šæŠ›å‡ºä¸€ä¸ªStopIterationé”™è¯¯ï¼Œå¹¶ä¸”returnçš„å€¼ä¼šæˆä¸ºerrorçš„valueå±æ€§ï¼š

### ç”Ÿæˆå™¨çš„`throw`å’Œ`close`æ–¹æ³•

ç”Ÿæˆå™¨æœ‰`throw`å’Œ`close`æ–¹æ³•ï¼Œå¹³æ—¶ç”¨çš„ä¸å¤šï¼Œä½†æ˜¯ç†è§£å®ƒä»¬å¯ä»¥æ›´æ·±åˆ»çš„ç†è§£ç”Ÿæˆå™¨ï¼Œä¸»è¦è¦æ³¨æ„ä¸¤ç‚¹ï¼š

- `generator.close()`ï¼šè‡´ä½¿ç”Ÿæˆå™¨åœ¨æš‚åœçš„`yield`è¡¨è¾¾å¼å¤„æŠ›å‡º`GeneratorExit`å¼‚å¸¸ã€‚å¦‚æœç”Ÿæˆå™¨æ²¡æœ‰å¤„ç†è¿™ä¸ªå¼‚å¸¸ï¼Œæˆ–è€…æŠ›å‡ºäº†`StopIteration`å¼‚å¸¸ï¼ˆé€šå¸¸æ˜¯æŒ‡è¿è¡Œåˆ°ç»“å°¾ï¼‰ï¼Œè°ƒç”¨æ–¹ä¸ä¼šæŠ¥é”™ã€‚å¦‚æœæ”¶åˆ°`GeneratorExit`å¼‚å¸¸ï¼Œç”Ÿæˆå™¨ä¸€å®šä¸èƒ½äº§å‡ºå€¼ï¼Œå¦åˆ™è§£é‡Šå™¨ä¼šæŠ›å‡º`RuntimeError`å¼‚å¸¸ã€‚ç”Ÿæˆå™¨æŠ›å‡ºçš„å…¶ä»–å¼‚å¸¸ä¼šå‘ä¸Šå†’æ³¡ï¼Œä¼ ç»™è°ƒç”¨æ–¹ã€‚

### æ·±å…¥ç†è§£`yield from` 

åœ¨ä»‹ç»`yield from`ä¹‹å‰ï¼Œå…ˆç¨å¾®å›é¡¾ä¸€ä¸‹æ™®é€šçš„ç”Ÿæˆå™¨ï¼ˆåç¨‹ï¼‰çš„å¤„ç†æœºåˆ¶ï¼Œå¦‚ä¸‹çš„ç”Ÿæˆå™¨ï¼š

````python
def c():
    yield 1
    yield 2
    return 42

g = c()
print(next(g)) # æ¿€æ´»ç”Ÿæˆå™¨
print(next(g))
try:
    next(g)
except StopIteration as e:
    print(e.value)
````

````python
def t():
    a = yield from c() # æ­¤æ—¶tè¢«æŒ‚èµ·ï¼Œæ‰§è¡Œcï¼Œå½“cæ‰§è¡Œå®Œæ¯•ï¼Œæ•æ‰cæŠ›å‡ºçš„StopIterationå®ä¾‹ï¼Œå¹¶å°†valueå±æ€§çš„å€¼èµ‹ç»™a
    yield a # è¿”å›tï¼Œä¾æ¬¡å‘ä¸‹æ‰§è¡Œï¼Œè¿™é‡Œå°†aè¿”å›

for i in t():
    print(i)
````

````python
class Score:
    def __init__(self, attrname):
        self.attrname = "_" + attrname
    
    def __get__(self, inst, cls):
        if inst is None:
            return self
        return getattr(inst, self.attrname)
    
    def __set__(self, inst, value):
        if not (0 <= value <= 100):
            raise ValueError("åˆ†æ•°ä¸èƒ½å¤§äº100å°äº0!")
        setattr(inst, self.attrname, value)


class Student:
    math = Score("math")
    english = Score("english")
````

ä¸Šé¢è¿™ç§åº”è¯¥æ˜¯æœ€ç®€å•çš„äº†ï¼Œä½†æ˜¯è¿˜æœ‰ä¸ªå°é—æ†¾å°±æ˜¯æ¯æ¬¡å®šä¹‰æè¿°ç¬¦ï¼Œéƒ½è¦ä¼ å…¥ä¸€ä¸ªå’Œå±æ€§åç›¸åŒçš„å­—ç¬¦ä¸²ï¼Œæ˜¾å¾—æœ‰ç‚¹å†—ä½™ï¼Œå¯ä»¥ç”¨å…ƒç±»è§£å†³ï¼š

````python
class Meta(type):
    def __new__(meta, clsname, bases, clsdict):
        for key, value in clsdict.items():
            if isinstance(value, Score):
                value.attrname = "_" + key
        return type.__new__(meta, clsname, bases, clsdict)


class Score:
    def __get__(self, inst, cls):
        if inst is None:
            return self
        return getattr(inst, self.attrname, 0)

    def __set__(self, inst, value):
        if not (0 <= value <= 100):
            raise ValueError("åˆ†æ•°ä¸èƒ½å¤§äº100å°äº0!")
        setattr(inst, self.attrname, value)


class Student(metaclass=Meta):
    math = Score()
    english = Score()
````

#### ä½¿ç”¨ä»…æœ‰`__get__`æ–¹æ³•çš„æè¿°ç¬¦è¿›è¡Œæ•°æ®çš„é«˜é€Ÿç¼“å­˜

å½“æŸä¸€ä¸ªå±æ€§éœ€è¦è¿›è¡Œå¤æ‚çš„è®¡ç®—ï¼Œå¯ä»¥ä½¿ç”¨ä»…æœ‰`__get__`æ–¹æ³•çš„æè¿°ç¬¦ï¼Œåœ¨ç¬¬ä¸€æ¬¡è¯»å–å±æ€§çš„æ—¶å€™è¿›è¡Œè®¡ç®—ï¼Œç„¶åå°†ç»“æœä¿å­˜ä¸ºå®ä¾‹çš„ä¸æè¿°ç¬¦åŒåçš„å±æ€§ï¼ŒåæœŸå°±ä¼šç›´æ¥è¯»å–å®ä¾‹å±æ€§ï¼Œä¸ä¼šå†è§¦å‘`__get__`æ–¹æ³•ï¼Œä»è€Œå®ç°äº†æ•°æ®çš„é«˜é€Ÿç¼“å­˜ï¼š

````python
class D:
    def __get__(self, inst, owner):
        if inst is None:
            return self  # å¦‚æœä»ç±»è®¿é—®æè¿°ç¬¦ï¼Œåˆ™è¿”å›æè¿°ç¬¦æœ¬èº«
        else:
            print("computing....")
            inst.data = 42  # 42ä»£è¡¨è®¡ç®—å‡ºçš„ç»“æœ
            return 42


class C:
    data = D()


c = C()
c.data  # ç¬¬ä¸€æ¬¡è¯»å–dataå±æ€§ï¼Œæ­¤æ—¶c.dataæ˜¯æè¿°ç¬¦ï¼Œè§¦å‘__get__æ–¹æ³•ï¼Œè¿›è¡Œè®¡ç®—ï¼Œè¿”å›ç»“æœï¼Œå¹¶ä¿å­˜ä¸ºå®ä¾‹çš„åŒåå±æ€§
c.data  # ç¬¬äºŒæ¬¡è¯»å–dataå±æ€§ï¼Œæ­¤æ—¶c.dataæ˜¯å®ä¾‹å±æ€§ï¼Œä¸ä¼šå†è§¦å‘__get__æ–¹æ³•ï¼Œä»è€Œå®ç°äº†æ•°æ®çš„é«˜é€Ÿç¼“å­˜
````

ä¸Šè¿°çš„åœºæ™¯éƒ½æ˜¯åˆ©ç”¨äº†æè¿°ç¬¦`__set__`å’Œ`__get__`æ–¹æ³•çš„ä¸å¯¹ç§°æ€§ï¼Œ`__set__`æ–¹æ³•æ˜¯è¦†ç›–å‹æ–¹æ³•ï¼Œå®ƒä¼šè¦†ç›–å®ä¾‹çš„åŒåå±æ€§ï¼Œæ„æ€æ˜¯å†™å…¥å±æ€§çš„æ—¶å€™ï¼Œå³ä½¿å®ä¾‹æœ‰å’Œæè¿°ç¬¦ç›¸åŒåå­—çš„å±æ€§ï¼Œå®ƒä¹Ÿä¼šè¿›è¡Œæ‹¦æˆªã€‚è€Œ`__get__`æ–¹æ³•æ˜¯éè¦†ç›–çš„ï¼Œå¦‚æœå®ä¾‹æœ‰å’Œæè¿°ç¬¦ç›¸åŒåç§°çš„å±æ€§ï¼Œåˆ™è¯»å–çš„æ—¶å€™ï¼Œä¸ä¼šå†è§¦å‘`__get__`æ–¹æ³•ï¼Œç›´æ¥è¯»å–å®ä¾‹çš„å±æ€§ã€‚

### è¦†ç›–å‹æè¿°ç¬¦å’Œéè¦†ç›–æè¿°ç¬¦

æˆ‘å¾ˆé•¿ä¸€æ®µæ—¶é—´éƒ½æé”™äº†ä¸€ä»¶äº‹æƒ…ï¼Œå°±æ˜¯ä¸€ç›´ä»¥ä¸ºè¦†ç›–ä¸è¦†ç›–æŒ‡çš„æ˜¯`__set__`å’Œ`__get__`æ–¹æ³•ï¼Œå³`__set__`æ–¹æ³•æ˜¯è¦†ç›–å‹çš„ï¼Œæ‰€æœ‰çš„å†™å…¥æ“ä½œéƒ½ä¼šæ‹¦æˆªï¼Œè€Œ`__get__`æ–¹æ³•æ˜¯éè¦†ç›–å‹çš„ï¼Œå¦‚æœå®ä¾‹æœ‰å’Œæè¿°ç¬¦ä¸€æ ·çš„åŒåå±æ€§ï¼Œåˆ™ä¼šä¼˜å…ˆè¯»å–å®ä¾‹çš„å±æ€§ï¼Œè€Œä¸ä¼šè§¦å‘æè¿°ç¬¦çš„`__get__`æ–¹æ³•ã€‚

æœ€è¿‘çªç„¶å‘ç°è‡ªå·±çš„ç†è§£å®Œå…¨é”™äº†ï¼Œå…¶å®æ˜¯æ ¹æ®æè¿°ç¬¦æœ‰è¿˜æ˜¯æ²¡æœ‰`__set__`æ–¹æ³•ï¼Œæ¥å†³å®šæ˜¯è¦†ç›–è¿˜æ˜¯ä¸è¦†ç›–ï¼Œå³ï¼Œå¦‚æœå®ä¾‹æœ‰åŒåå±æ€§ï¼Œä½†æ˜¯åªè¦æè¿°ç¬¦æœ‰`__set__`æ–¹æ³•ï¼Œé‚£ä¹ˆè¿™ä¸ªæè¿°ç¬¦æ˜¯è¦†ç›–å‹çš„ï¼Œä¸ç®¡æ˜¯è¯»è¿˜æ˜¯å†™ï¼Œéƒ½ä¼šè¿›è¡Œæ‹¦æˆªã€‚è€Œå¦‚æœæ²¡æœ‰`__set__`æ–¹æ³•ï¼Œé‚£ä¹ˆå°±æ˜¯éè¦†ç›–å‹çš„æè¿°ç¬¦ï¼Œå¦‚æœæœ‰åŒåå±æ€§ï¼Œä¼šä¼˜å…ˆè¯»å–å®ä¾‹çš„å±æ€§ã€‚

````python
class D:
    def __set__(self, inst, value):
        print("setting")
        inst.__dict__["data"] = value

    def __get__(self, inst, cls):
        print("getting")
        if inst is None:
            return self
        return inst.__dict__["data"]


class C:
    data = D()


c = C()
c.data = 42
print(c.__dict__)  # å®ä¾‹å·²ç»æœ‰äº†dataå±æ€§ï¼Œä½†æ˜¯å› ä¸ºæè¿°ç¬¦æœ‰`__set__`æ–¹æ³•ï¼Œæ­¤æ—¶è¯»å–çš„æ—¶å€™ä»ç„¶ä¼šè¿›è¡Œæ‹¦æˆªï¼Œè§¦å‘__get__æ–¹æ³•
print(c.data)
````

### ä½¿ç”¨æè¿°ç¬¦åšæ–¹æ³•çš„è£…é¥°å™¨

æˆ‘ä»¬å¯ä»¥ç”¨å‡½æ•°ï¼Œå’Œç±»ä½œè£…é¥°å™¨ï¼Œä½†æ˜¯ï¼Œåœ¨è£…é¥°æ–¹æ³•çš„æ—¶å€™ï¼Œç±»ä½œä¸ºè£…é¥°å™¨å°±ä¸å¥½ç”¨äº†ï¼Œå¦‚ä¸‹ï¼š

````python
class Wrapper:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        print("in wrapper")
        return self.func(*args, **kwargs)


@Wrapper
def func(a, b):
    return a + b


func(2, 3)
````

ä½¿ç”¨`Wrapper`ç±»åŒ…è£…æ–¹æ³•æ—¶ï¼Œä¼šæŠ¥é”™ï¼Œå› ä¸ºæ­¤æ—¶`c.func`è¿”å›çš„æ˜¯`Wrapper`çš„å®ä¾‹ï¼Œç„¶åå†è°ƒç”¨`Wrapper`å®ä¾‹çš„`__call__`æ–¹æ³•ï¼Œè¿™ä¸ªè¿‡ç¨‹ä¸¢å¤±äº†`C`çš„`self`å®ä¾‹çš„ä¼ é€’ï¼š

````python
class C:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    @Wrapper
    def func(self):
        return self.a + self.b


c = C(2, 3)
try:
    c.func()
except TypeError as e:
    print(e)
````

````python
class Wrapper:
    def __init__(self, func):
        self.func = func
        
    def __get__(self, inst, cls):
        def wrapper(*args, **kwargs):           
            print("in wrapper")
            return self.func(inst, *args, **kwargs)
        return wrapper


class C:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    @Wrapper
    def func(self):
        return self.a + self.b

c = C(2, 3)
c.func()
````

````python
class MMCls(type):
    def __call__(metacls, clsname, base, attrs):
        print("enter mmclass __call__")
        cls = super().__call__(clsname, base, attrs)
        print("exit mmclass __call__")
        return cls


class MCls(type, metaclass=MMCls):
    def __new__(metacls, clsname, base, attrs):
        print("enter mclass __new__")
        cls = super().__new__(metacls, clsname, base, attrs)
        print("exit mclass __new__")
        return cls

    def __init__(cls, clsname, base, attrs):
        print("enter mclass __init__")
        super().__init__(clsname)
        print("exit mclass __init__")


class Cls(metaclass=MCls):
    pass
````

ç±»åˆ›å»ºå®ä¾‹çš„è¿‡ç¨‹ä¹Ÿæ˜¯ä¸€æ ·çš„ï¼Œå…ˆè°ƒç”¨å…ƒç±»çš„__call__æ–¹æ³•ï¼Œåœ¨å…ƒç±»çš„__call__æ–¹æ³•ä¸­ï¼Œä¾æ¬¡è°ƒç”¨ç±»çš„__new__å’Œ__init__æ–¹æ³•ï¼Œæœ€åè¿”å›ä¸€ä¸ªå®ä¾‹ã€‚æ¯”å¦‚ï¼Œå½“å®ç°å•ä¾‹æ¨¡å¼çš„æ—¶å€™ï¼Œå°±å¯ä»¥é€šè¿‡å…ƒç±»çš„__call__æ–¹æ³•æ¥å®ç°ï¼Œè¿™æ ·çš„å¥½å¤„æ˜¯ï¼Œä¸ä¼šæ¯æ¬¡åˆ›å»ºå®ä¾‹éƒ½ä¼šè°ƒç”¨__init__æ–¹æ³•ï¼Œå¦‚ä¸‹ï¼š

````python
class C:
    _inst = None
    
    def __new__(cls, *args, **kwargs):
        if C._inst == None:
            inst = super().__new__(cls, *args, **kwargs)
            C._inst = inst
            return inst
        else:
            return C._inst
        
    def __init__(self):
        print("create a C instance")
        
c1 = C()
c2 = C()
c1 is c2
````

å¯è§ï¼Œè™½ç„¶ä¸¤æ¬¡åˆ›å»ºçš„æ˜¯åŒä¸€ä¸ªå®ä¾‹ï¼Œä½†æ˜¯æ¯æ¬¡è°ƒç”¨C()ï¼Œéƒ½ä¼šæ‰§è¡Œä¸€æ¬¡initå‡½æ•°ã€‚

````python
class MetaC(type): 
    def __init__(cls, *args, **kwargs):
        cls._inst = None
        super().__init__(*args, **kwargs)
        
    def __call__(cls, *args, **kwargs):
        if cls._inst == None:
            inst = super().__call__(*args, **kwargs)
            cls._inst = inst
            return inst
        else:
            return cls._inst
        
class C(metaclass=MetaC):
    def __init__(self):
        print("create a C instance")
        
c1 = C()
c2 = C()
c1 is c2
````

#### `concurrent.futures`çš„ç»†èŠ‚

ä½¿ç”¨å¤šçº¿ç¨‹å’Œå¤šè¿›ç¨‹æœ€ç®€å•çš„æ–¹æ³•ï¼Œå°±æ˜¯ä½¿ç”¨å†…éƒ¨åº“çš„`concurrent.futures`æ¨¡å—ï¼Œå®ƒæ˜¯å¯¹å†…éƒ¨çš„`threading`å’Œ`multiprocessing`åº“çš„ä¸Šå±‚å°è£…ï¼Œåˆ†åˆ«å¯¹åº”`futures.ThreadPoolExecutor()`å’Œ`Futures.ProcessPoolExecutor()`ä¸¤ä¸ªæ–¹æ³•ã€‚ä¸¤ç§æ–¹æ³•éƒ½è¿”å›ä¸€ä¸ª`Executor`æ‰§è¡Œå™¨ï¼Œä½¿ç”¨çš„æ—¶å€™ï¼Œä¸»è¦æœ‰ä¸¤ç§æ–¹å¼ï¼Œä¸€ç§æ˜¯ä½¿ç”¨`executor`çš„`map`æ–¹æ³•ï¼Œä¸€ç§æ˜¯ä½¿ç”¨`executor`çš„`submit`æ­é…`as_completed`æ–¹æ³•ã€‚
1. `map`æ–¹æ³•å¦‚ä¸‹å‡ ç‚¹ä¸»è¦æ³¨æ„ï¼š
    - `map`æ–¹æ³•ç›´æ¥è¿”å›ä¸€ä¸ªç»“æœçš„ç”Ÿæˆå™¨ï¼Œç”Ÿæˆå™¨å†…éƒ¨å®é™…ä¸Šæ˜¯ä¸€ä¸ªæœŸç¨‹ï¼ˆ`Future`ï¼‰çš„åˆ—è¡¨ï¼Œè¿­ä»£ç”Ÿæˆå™¨æ—¶ï¼Œå®é™…ä¸Šä¾æ¬¡è°ƒç”¨æœŸç¨‹ï¼ˆ`Future`ï¼‰çš„`result()`çš„æ–¹æ³•ã€‚å¦‚æœçº¿ç¨‹è¿˜æ²¡æœ‰æ‰§è¡Œå®Œï¼Œä¼šåœ¨è¿™é‡Œé˜»å¡ï¼Œç­‰å¾…`result()`è¿”å›ç»“æœï¼Œå¦‚æœå‘ç”Ÿå¼‚å¸¸ï¼Œä¹Ÿä¼šåœ¨è¿™é‡ŒæŠ›å‡ºï¼ŒåŒæ—¶ç»“æŸï¼Œè¿”å›ä¸»çº¿ç¨‹ã€‚å› æ­¤æœ€åçš„ç»“æœå’Œä¼ å…¥å‚æ•°é¡ºåºæ˜¯ä¸€æ ·çš„ã€‚

````python
from concurrent import futures
from time import sleep
from datetime import datetime
````

````python
def now():
    now = datetime.now()
    return now.strftime("%Y-%m-%d %H:%M:%S")
````

````python
def func1():
    print(f"func1 start, time is {now()}")
    sleep(4)
    return f"func1 end, time is {now()}"

def func2():
    print(f"func2 start, time is {now()}")
    sleep(2)
    return f"func2 end, time is {now()}"

def func3():
    print(f"func3 start, time is {now()}")
    sleep(3)
    raise TypeError("func3 error")
    
def func4():
    print(f"func4 start, time is {now()}")
    sleep(5)
    return f"func4 end, time is {now()}"
````

çº¿ç¨‹æ ¹æœ¬ä¸Šè¿˜æ˜¯æ—¶é—´æ’æœŸï¼Œå¹¶éçœŸæ­£çš„å¹¶å‘ï¼Œé€šè¿‡ä»¥ä¸‹ä¸¤ä¸ªä¾‹å­ä»”ç»†ä½“ä¼šæ¯ä¸ªæ—¶é—´ç‚¹ï¼š

````python
with futures.ThreadPoolExecutor(max_workers=3) as executor:
    # ç«‹å³å¯¹æ‰€æœ‰çº¿ç¨‹è¿›è¡Œæ’æœŸï¼Œå¼€å§‹å¹¶å‘æ‰§è¡Œï¼Œè¿”å›ä¸€ä¸ªç”Ÿæˆå™¨ï¼Œä¿ç•™äº†åŠ å…¥æ—¶çš„é¡ºåº
    results = executor.map(lambda func:func(), [func1, func2, func3, func4]) 
    # å¹¶å‘å¯åŠ¨3ä¸ªçº¿ç¨‹ä»¥åï¼Œå›åˆ°ä¸»çº¿ç¨‹å‘ä¸‹æ‰§è¡Œ
    sleep(1)
    print(f"waiting...time is {now()}") 
    # 2ç§’é’Ÿåï¼Œfunc1çº¿ç¨‹ç»“æŸï¼Œå¯åŠ¨func4ï¼Œç„¶ååˆå›åˆ°ä¸»çº¿ç¨‹å‘ä¸‹æ‰§è¡Œ
    try:
        # è¿™é‡Œéšå¼ä¾æ¬¡è°ƒç”¨åº•å±‚futureçš„result()æ–¹æ³•ï¼Œå¦‚æœçº¿ç¨‹è¿˜æ²¡æœ‰æ‰§è¡Œå®Œï¼Œä¼šåœ¨è¿™é‡Œé˜»å¡ï¼Œå¼‚å¸¸ä¹Ÿåœ¨è¿™é‡ŒæŠ›å‡º
        for r in results:  
            print(r)
    # æ•æ‰åˆ°func3æŠ›å‡ºçš„é”™è¯¯ï¼Œåé¢çš„çº¿ç¨‹ä¼šç»“æŸè¿è¡Œ
    except TypeError as e: 
        print(e)
````

````python
with futures.ThreadPoolExecutor(max_workers=3) as executor:
    results = executor.map(lambda func:func(), [func1, func2, func3, func4]) 
    sleep(1)
    print(f"waiting...time is {now()}")
# å°±ç®—ä¸è°ƒç”¨æœŸç¨‹çš„resultæ–¹æ³•ï¼Œä¹Ÿä¼šåœ¨withä»£ç å—çš„æœ€åç­‰å¾…æ‰€æœ‰çº¿ç¨‹æ‰§è¡Œå®Œæ¯•
print(f"out with block, time is {now()}")
````

````python
with futures.ThreadPoolExecutor(max_workers=3) as executor:
    future1 = executor.submit(func1)  # è¿”å›ä¸€ä¸ªå·²æ’æœŸçš„æœŸç¨‹ï¼Œåœ¨è¿™é‡Œä¼šç«‹å³æ‰§è¡Œ
    future2 = executor.submit(func2)
    future3 = executor.submit(func3)
    sleep(1)  # åŒæ ·çš„ï¼Œå¯åŠ¨ä»¥åç«‹å³å›åˆ°ä¸»çº¿ç¨‹å‘ä¸‹æ‰§è¡Œ
    print(f"waiting...time is {now()}")
    # è¿”å›func4çš„æœŸç¨‹ï¼Œä½†æ˜¯è¿™é‡Œä¸ä¼šé©¬ä¸Šæ‰§è¡Œï¼Œå› ä¸ºçº¿ç¨‹æ± æ˜¯æ»¡çš„
    future4 = executor.submit(func4)
    # è¿™é‡Œå¹¶ä¸ä¼šé˜»å¡ï¼Œåªæ˜¯è¿”å›ä¸€ä¸ªæœŸç¨‹æ„æˆçš„ç”Ÿæˆå™¨
    done_iter = futures.as_completed([future1, future2, future3, future4])
    print(f"after as_completed...time is {now()}")
    # æ­¤æ—¶func2æ‰§è¡Œå®Œï¼Œçº¿ç¨‹æ± ç©ºäº†ä¸€ä¸ªï¼Œå¼€å§‹å¯åŠ¨func4çš„çº¿ç¨‹,ç„¶åé©¬ä¸Šåˆå›åˆ°ä¸»çº¿ç¨‹
    try:
        # ä¸€ä¸ªçº¿ç¨‹ç»“æŸæ‰ä¼šè¿›å…¥ç”Ÿæˆå™¨å¹¶è¿”å›ï¼Œå¦‚æœçº¿ç¨‹æ²¡æœ‰æ‰§è¡Œå®Œå°±é˜»å¡ç­‰å¾…ï¼Œæ‰€ä»¥å“ªä¸ªå…ˆæ‰§è¡Œå®Œå“ªä¸ªå…ˆè¿”å›ï¼Œä¸ä¼šä¿ç•™é¡ºåº
        for f in done_iter: 
            print(f"waiting...time is {now()}")
            # è¿™é‡Œä¸ä¼šé˜»å¡ï¼Œå› ä¸ºæ­¤æ—¶çš„futureæ˜¯å·²ç»æ‰§è¡Œå®Œæ¯•çš„futureï¼Œæ³¨æ„ï¼šå¦‚æœæœ‰å¼‚å¸¸ï¼Œä¼šåœ¨è¿™é‡ŒæŠ›å‡ºã€‚
            print(f.result())
    # å¦‚æœåœ¨forå¾ªç¯å¤–å±‚æ•è·é”™è¯¯ï¼Œåˆ™ä¼šå’Œmapä¸€æ ·ï¼Œç›´æ¥ç»“æŸ
    except TypeError as e:
        print(e)
````

3. æœŸç¨‹çš„å…¶å®ƒæ–¹æ³•ï¼š
    - `done()`ï¼šæŸ¥çœ‹æœŸç¨‹æ˜¯å¦æ‰§è¡Œå®Œæ¯•ã€‚
    - `add_done_callback()`ï¼šç»™æœŸç¨‹æ·»åŠ ä¸€ä¸ªå›è°ƒå‡½æ•°ã€‚æ³¨æ„ï¼šè¿™ä¸ªå›è°ƒå‡½æ•°åªæ¥å—ä¸€ä¸ªå‚æ•°å°±æ˜¯æœŸç¨‹ã€‚
    - `exception()`ï¼šå’Œ`result()`ç±»ä¼¼ï¼Œé˜»å¡ä¸»çº¿ç¨‹ï¼Œå¦‚æœæœ‰å¼‚å¸¸åˆ™è¿”å›ï¼Œå¦‚æœè¿è¡Œæ­£å¸¸åˆ™è¿”å›Noneã€‚

````python
from functools import partial


def t():
    print(f"t started, time is {now()}")
    sleep(1)
    return 42


def cbfunc(f, func):
    if f.done():
        print(f"{func.__name__} end, time is {now()}")


with futures.ThreadPoolExecutor(max_workers=3) as executor:
    f = executor.submit(t)
    f.add_done_callback(partial(cbfunc, func=t))
    f.result()
````

````python
# 2020å¹´12æœˆ28æ—¥è¡¥å……ï¼šèƒ½å¦ç»“æŸæ‰€æœ‰çš„çº¿ç¨‹ï¼Ÿ

from queue import Queue, LifoQueue, PriorityQueue


class IterQueueMixin:
    SIG = object()

    def close(self, n_workers=1):
        for i in range(n_workers):
            self.put(self.SIG)

    def __iter__(self):
        while True:
            item = self.get()
            try:
                if item is self.SIG:
                    return  # ç”Ÿæˆå™¨ä¸­returnä¼šæŠ›å‡ºStopIterationé”™è¯¯
                yield item
            finally:
                self.task_done()
                

class IterQueue(IterQueueMixin, Queue):
    pass


class IterLifoQueue(IterQueueMixin, LifoQueue):
    pass
````

```python
import asyncio
import time


async def c1():
```

````python
import asyncio
import time


async def c():
    print(f"c started at {time.strftime('%X')}")
    await asyncio.sleep(3)
    print(f"c ended at {time.strftime('%X')}")
    return 42


loop = asyncio.get_running_loop()
t = loop.create_task(c())
````

```python
with ExitStack() as stack:
    files = (stack.enter_context(open(file, "rt")) for file in ['file1.txt', 'file2.txt'])
```

ExitStackçš„å‡ ä¸ªæ–¹æ³•éƒ½ä¸æ˜¯å¾ˆå¥½ç†è§£ï¼Œæ¥ä¸‹æ¥ä¸€ä¸€è¿›è¡Œè§£é‡Šï¼š
1. `stack.enter_context(open(file, "rt")`ç›¸å½“äºè°ƒç”¨äº†`with open(file, "rt") as f`ï¼Œè¿”å›fã€‚åŒæ—¶å°†`open(file, "rt")`ä¸Šä¸‹æ–‡ç®¡ç†å™¨çš„`__exit__`æ–¹æ³•å‹å…¥ä¸€ä¸ªå †æ ˆã€‚å½“`with ExitStack()`ç»“æŸçš„æ—¶å€™ï¼Œä¼šä¾æ¬¡å¼¹å‡ºå †æ ˆçš„`__exit__`æ–¹æ³•å¹¶è¿è¡Œã€‚æ³¨æ„å®ƒçš„é”™è¯¯å¤„ç†çš„æ—¶é—´ç‚¹ï¼Œå½“é”™è¯¯å‘ç”Ÿåœ¨æœ€åæ—¶ï¼Œæ€»æ˜¯è¢«æœ€ååŠ å…¥å †æ ˆçš„`__exit__`å‡½æ•°æ•è·ã€‚å¦‚ä¸‹ï¼š

````python
class CM:
    def __init__(self, name):
        self.name = name
        
    def __enter__(self):
        print(f'enter {self.name}')
        return self

    def __exit__(self, exc_type, exc_value, exc_tb):
        print(f'exit {self.name}')
        if exc_type:
            print(f"catch {exc_value}")
        return True


def es_demo1():
    with ExitStack() as stack:
        stack.enter_context(CM('cm1'))         
        stack.enter_context(CM('cm2'))     
        raise RuntimeError('error occured!') # æ•è·åˆ°é”™è¯¯ï¼Œä¼šä¾æ¬¡æ‰§è¡Œstacké‡Œé¢çš„__exit__

def es_demo2():
    with ExitStack() as stack:
        stack.enter_context(CM('cm1'))    
        raise RuntimeError('error occured!') # å¼‚å¸¸åé¢çš„è¯­å¥ä¸ä¼šæ‰§è¡Œ
        stack.enter_context(CM('cm2'))     
                
es_demo1()
print()
es_demo2()
````

2. æŸäº›æƒ…å†µä¸‹ï¼Œéœ€è¦é•¿æ—¶é—´ä¿æŒæŸä¸ªèµ„æºï¼Œæ¯”å¦‚ä¸€ä¸ªç½‘ç»œè¿æ¥ï¼Œè¿™ä¸ªç½‘ç»œè¿æ¥çš„å­˜åœ¨æ—¶é—´ç”šè‡³è¦æ¯”ExitStackçš„ä¸Šä¸‹æ–‡ç®¡ç†å™¨å­˜åœ¨çš„æ—¶é—´è¿˜è¦é•¿ï¼Œå½“withç»“æŸçš„æ—¶å€™ï¼Œå¹¶ä¸æƒ³è¿›è¡Œæ¸…ç†ï¼Œæ­¤æ—¶å¯ä»¥ä½¿ç”¨`pop_all()`å’Œ`close`ï¼Œ`pop_all()`ä¼šè¿”å›ä¸€ä¸ªå †æ ˆï¼Œå…¶ä¸­å¡«å……äº†åŒæ ·çš„ä¸Šä¸‹æ–‡ç®¡ç†å™¨å’Œå›è°ƒï¼Œå¯ä»¥ç†è§£æˆåŸæ¥çš„ä¸€ä¸ªå‰¯æœ¬ï¼ŒåŒæ—¶åŸæ¥çš„å †æ ˆæ¸…ç©ºã€‚å¦‚ä¸‹ï¼š

````python
def es_demo3():
    with ExitStack() as stack:
        stack.enter_context(CM('cm1'))         
        stack.enter_context(CM('cm2'))     
        raise RuntimeError("error occurred!") 
        # å¦‚æœåœ¨ExitStacké‡Œé¢å‘ç”Ÿäº†å¼‚å¸¸ï¼Œåˆ™ä¼šç›´æ¥ç”±__exit__å¤„ç†ï¼Œä¸ä¼šæ‰§è¡Œä¸‹é¢çš„è¯­å¥ï¼Œå¹¶ä¸”æœ€ç»ˆè¿”å›None
        print("after raise exception!")
        return stack.pop_all()

stack = es_demo3()
````

````python
def es_demo4():
    with ExitStack() as stack:
        stack.enter_context(CM('cm1'))         
        stack.enter_context(CM('cm2'))     
        return stack.pop_all()
````

```python
class closing(object):
```
