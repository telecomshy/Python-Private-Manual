# Python爬虫私房手册-瑞士军刀Selenium 

`requests`只能发送`get`、`post`请求，但是现在很多网页都是通过`javascript`动态生成的，此时就需要`selenium`出马了。

## 准备工作 

安装`selenium`以后，还需要安装浏览器驱动，因为`selenium`本质上就像一个按键精灵一样。`selenium`支持各大浏览器，只要安装对应的驱动，注意启动要和浏览器的版本对应。
- [firefox驱动下载地址](https://github.com/mozilla/geckodriver/releases/)
- [chrome驱动下载地址](https://npm.taobao.org/mirrors/chromedriver/)

下载好了以后放到`python`的根目录，比如我装的是Anaconda,就放到`D:\programs\Anaconda\`文件夹下，其实只要是在系统的`path`环境变量路径下都可以，注意是系统的`path`，而不是`python`的`path`路径。也可以通过`webdriver.Chrome()`的`excutable_path`路径指定。

## 注意事项

### 无界面设置

默认情况下会弹出浏览器的窗口，如果不希望有界面，可以添加选项（网上的搜索出来的一些方法版本比较老，以下方法的`selenium`版本为3.141.0）：

````python
from selenium import webdriver

opt = webdriver.ChromeOptions()
opt.add_argument("--headless")

driver = webdriver.Chrome(options=opt)
driver.get("https://www.sina.com.cn")
driver.close()
````

### 查找元素

查找元素有两种方法：
1. 使用`find_element()`方法。
2. 使用`find_element_by_`方法。

`find_element_by_`指根据`id`，`name`等获取元素，主要有以下几种：


`find_element`更灵活，接受2个参数，查找方法`By`和具体的值，如下：
```
from selenium.webdriver.common.by import By

driver.find_element(By.ID, "q")
```

注意，以上方法查找出来的都是单个元素，如果有多个符合条件，只会返回第一个，如果需要返回多个元素，使用`find_elements`方法。`find_element`返回的是单个`WebElement`对象，`find_elements`返回的是`WebElement`对象的列表，接下来可以调用对象的各种方法，比如`send_keys()`发送文字，`clear()`清空文字，`click()`点击等等。
