### 概率分布

#### 几何分布的期望到底代表什么？

几何分布看起来最简单，但是仔细思考却最感觉各种违和，其中之一就是几何分布的期望，$E(x)=\frac{1}{p}$，比如每一次的成功率为0.2，可以得出期望值为5，直觉应该是第5次成功，因为几何分布求得就是第$n$次成功的概率，其随机变量为$n$。但是又觉得违和，应该这样思考：

期望指$n$无穷大，实验无数次，最终的平均值，即可能第1次成功，第3次成功，第10次成功。。。所有这些次数加起来除以n，正好等于5，即如果实验无数次的话，平均值为5，即第5次成功，这个5仅仅只是平均值，并不是说我们每一次实验第5次获得成功的几率最大，单次实验的成功率都是一样的，但是如果头几次一直失败，那么我们可以期望在第5次成功，当然失败次数可以超过5次，但是越往后，这种可能性越小。

## 统计

### 深入理解$R^2$决定系数

先上个图：

基本公式是：

$R^2=1-\frac{SSE}{SST}$

其中：
- $SSE$为残差平方和，即所有点的实际值减去预测的值的平方和
- $SST$为总的平方和，为所有点的实际值减去均值的平方和。
- 另外还有一个$SSR$，为所有点的预测值减去均值的平方和。

理解的关键是，$\frac{SSE}{SST}$是表示$y$的总波动中有多少不能被$x$解释，$1-\frac{SSE}{SST}$则表示有多少可以被$x$解释。

设想，如果能够百分百拟合，那么$SSE$为0，那么到均值的误差全部都能被$x$解释，也就是只要知道$x$，就可以把误差(注意这个误差是到均值的误差)算出来，此时$R^2$为1。如果不在拟合的线上，那么$SSE$这部分误差则是$x$解释不了的，即无法通过$x$计算出来。

仔细思考的话，仍然有疑惑，为什么$\frac{SSE}{SST}$就表示不能被$x$解释的呢，最后终于明白了：

因为$SST=SSR+SSE$，因此，公式也等价于$R^2=\frac{SSR}{SST}$，因为$SSR$的点都在预测线上，所以$SSR$表示可以被$x$解释的误差，那么，$SSE$自然就表示不能被$x$解释的误差了。为什么$SST=SSR+SSE$呢，推导如下：

其中根据最小二乘法，代价函数$SSE=\Sigma(y-\hat{y})^2$，对$a$求偏导得出$\Sigma(y-a-bx)x$为0，对$b$求偏导可知$\Sigma(y-a-bx)$为0。
