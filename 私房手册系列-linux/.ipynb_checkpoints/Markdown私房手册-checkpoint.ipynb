## 如何在列表里面嵌入代码？

不同的编辑器可能表现不同，常见的问题是：当在列表里面写入代码块，后面的列表会重新计数。解决方法是：在代码块前空一行，然后代码块前保留
4\*n个空格，n代表列表的层级，像这样（jupyter正常写就可以，可以自动识别）：
1. item1
2. item2

    ```Python
    print('hello world')
    ```
3. item3
